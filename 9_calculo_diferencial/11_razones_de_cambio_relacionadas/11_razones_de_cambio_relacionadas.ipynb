Este es un excelente problema de **razones de cambio relacionadas**. Para resolverlo, primero debemos modelar la geometría del cono truncado (o tronco de cono) y luego implementar la animación que visualice cómo sube el nivel del agua.

### Análisis del Problema

* **Datos constantes:**
    * Radio inferior ($r_1$): $20\text{ cm}$
    * Radio superior ($R$): $40\text{ cm}$
    * Altura total del tanque ($H$): $80\text{ cm}$
    * Razón de llenado ($\frac{dV}{dt}$): $2000\text{ cm}^3/\text{min}$ ($2\text{ litros/min}$)
* **Relación de semejanza:** El radio $r$ en cualquier altura $h$ sigue una progresión lineal:
    $$r(h) = r_1 + \left(\frac{R - r_1}{H}\right)h = 20 + \frac{20}{80}h = 20 + \frac{1}{4}h$$
* **Volumen del agua:** El volumen de un tronco de cono de altura $h$ es:
    $$V = \frac{\pi h}{3}(r_1^2 + r_1r + r^2)$$

### Código para la Animación con `ipywidgets`

Este código generará un gráfico 2D del tanque y una representación azul que sube según el volumen de agua ingresado. 

```python


````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def solve_filling_frustum():
    # Parámetros del tanque (en cm)
    r_bottom = 20
    r_top = 40
    H = 80
    dVdt = 2000  # cm^3/min (2 Litros/min)

    def get_volume(h):
        """Calcula el volumen de agua para una altura h dada."""
        r_h = r_bottom + (r_top - r_bottom) * (h / H)
        return (np.pi * h / 3) * (r_bottom**2 + r_bottom * r_h + r_h**2)

    def plot_tank(time):
        # Calcular el volumen actual basado en el tiempo
        current_volume = dVdt * time
        
        # Encontrar la altura h correspondiente al volumen mediante búsqueda (0 a 80)
        # Para la animación, usamos una aproximación simple
        hs = np.linspace(0, H, 500)
        vs = [get_volume(h) for h in hs]
        h_current = np.interp(current_volume, vs, hs)
        r_curr = r_bottom + (r_top - r_bottom) * (h_current / H)

        # Configuración del gráfico
        fig, ax = plt.subplots(figsize=(6, 8))
        
        # Dibujar contorno del tanque
        tank_x = [r_bottom, r_top, -r_top, -r_bottom, r_bottom]
        tank_y = [0, H, H, 0, 0]
        ax.plot(tank_x, tank_y, color='black', lw=3)
        
        # Dibujar el agua (relleno azul)
        water_x = [r_bottom, r_curr, -r_curr, -r_bottom, r_bottom]
        water_y = [0, h_current, h_current, 0, 0]
        ax.fill(water_x, water_y, color='skyblue', alpha=0.7, label='Agua')

        # Cálculo de la velocidad instantánea dh/dt en ese punto
        # Usando la fórmula: dh/dt = (dV/dt) / (pi * r^2)
        dhdt = dVdt / (np.pi * r_curr**2)

        # Detalles estéticos
        ax.set_xlim(-50, 50)
        ax.set_ylim(-5, 90)
        ax.set_aspect('equal')
        ax.set_title(f"Llenado del Tanque\nTiempo: {time:.1f} min")
        ax.set_xlabel("Radio (cm)")
        ax.set_ylabel("Altura (cm)")
        
        # Anotaciones de datos en tiempo real
        plt.text(45, h_current, f"h = {h_current:.2f} cm", verticalalignment='center')
        plt.text(-45, 85, f"v_ascenso (dh/dt) ≈ {dhdt:.4f} cm/min", color='blue')
        
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.show()

    # El tiempo máximo es el volumen total / dVdt
    v_total = get_volume(H)
    max_time = v_total / dVdt

    interact(plot_tank, time=FloatSlider(min=0, max=max_time, step=0.5, value=0, description='Minutos:'))

solve_filling_frustum()
````

# La epistemologís del concepto de derivada  

Isaac Newton no “inventó el cálculo diferencial” por curiosidad abstracta únicamente, sino por una necesidad muy concreta: **describir y entender el cambio en la naturaleza**.

En su época (siglo XVII), había problemas que las matemáticas existentes no podían resolver bien:

* ¿Cómo calcular la **velocidad instantánea** de un objeto en movimiento?
* ¿Cómo describir trayectorias curvas, como las de los planetas?
* ¿Cómo modelar fenómenos físicos que cambian continuamente?



### 🔑 La idea clave

Newton desarrolló el cálculo diferencial para estudiar **cómo cambian las cantidades en el tiempo o en el espacio**.

Esto se conecta con lo que hoy llamamos:

* Derivadas → tasas de cambio (por ejemplo, velocidad)
* Pendientes de curvas → cómo se comporta una función en un punto

### 🌍 Aplicación principal: la física

Newton necesitaba estas herramientas para formular sus leyes, especialmente:

* Las leyes del movimiento
* La ley de la gravitación universal

Sin cálculo, no habría podido expresar con precisión cómo:

* Un objeto acelera
* Un planeta orbita
* Una fuerza produce movimiento

### ⚡ Ejemplo intuitivo

Si tienes la posición de un objeto ( x(t) ), Newton quería responder:

* ¿Qué tan rápido se mueve en un instante? → derivada → velocidad
* ¿Cómo cambia esa velocidad? → segunda derivada → aceleración

### 🧠 Un detalle importante



Al mismo tiempo, de forma independiente, Gottfried Wilhelm Leibniz desarrolló una versión del cálculo (con una notación diferente, que es la que usamos hoy en día).

---

### 🎯 En resumen

Newton desarrolló el cálculo diferencial para:

* Estudiar el **cambio continuo**
* Resolver problemas de **movimiento y física**
* Crear un lenguaje matemático capaz de describir la naturaleza
