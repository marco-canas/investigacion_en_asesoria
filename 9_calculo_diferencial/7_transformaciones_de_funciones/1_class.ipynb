Dame una función de usuario de Python, para tomar asistencia en el espacio de asesoría, donde la función: consigne los datos tomados de asistencia en un archivo .xlsx. Te pido que la función le asigne un identificador corto a cada estudiante asistente a la asesoría, para que cada vez que asista de nuevo a la asesoría, no sea necesario preguntarle todos los datos sino solo ingresar el identificador o nombre corto para el estudiante y la dificultad o solicitud que lo trae al espacio de asesoría. Dame el diseño  de esta función, de tal manera que ella consigne, nombre del estudiante, correo o email, programa académico y dificultad o solicitud que lo trae al espacio a asesoría. 

````python
# Toma de asistencia rápida al espacio de asesoría  

from openpyxl import Workbook, load_workbook
import os  

def registrar_asistencia():
    """
    Función para registrar asistencia en un espacio de asesoría.
    Guarda nombre, correo, programa académico y dificultad/solicitud.
    El archivo se guarda y actualiza en la carpeta especificada.
    """

    # Ruta donde se guardará el archivo
    carpeta = r"C:\Users\marco\OneDrive - Universidad de Antioquia\Documentos\1_administracion\informes_jefe\renovacion_contrato_ocasional\asistencia_asesoria"
    archivo = os.path.join(carpeta, "asistencia.xlsx")

    # Crear carpeta si no existe
    os.makedirs(carpeta, exist_ok=True)

    # Verificar si existe el archivo, si no lo crea
    if os.path.exists(archivo):
        wb = load_workbook(archivo)
        ws = wb.active
    else:
        wb = Workbook()
        ws = wb.active
        ws.append(["ID", "Nombre", "Correo", "Programa", "Dificultad/Solicitud"])
    
    # Diccionario de estudiantes ya registrados (ID como clave)
    estudiantes = {}
    for row in ws.iter_rows(min_row=2, values_only=True):
        estudiantes[row[0]] = {"nombre": row[1], "correo": row[2], "programa": row[3]}
    
    # Pedir identificador o nombre corto
    identificador = input("Ingrese su identificador corto o nuevo ID: ").strip()
    
    if identificador in estudiantes:
        print(f"Bienvenido de nuevo {estudiantes[identificador]['nombre']}")
        dificultad = input("Ingrese la dificultad o solicitud que lo trae: ")
        ws.append([identificador, estudiantes[identificador]['nombre'], 
                   estudiantes[identificador]['correo'], 
                   estudiantes[identificador]['programa'], 
                   dificultad])
    else:
        print("Nuevo estudiante, por favor ingrese sus datos completos.")
        nombre = input("Nombre completo: ")
        correo = input("Correo electrónico: ")
        programa = input("Programa académico: ")
        dificultad = input("Ingrese la dificultad o solicitud que lo trae: ")
        
        estudiantes[identificador] = {"nombre": nombre, "correo": correo, "programa": programa}
        ws.append([identificador, nombre, correo, programa, dificultad])
    
    # Guardar cambios en la ruta indicada
    wb.save(archivo)
    print(f"Asistencia registrada en {archivo}")

````

Vamos a representar la **secuencia de transformaciones** desde
[
f(x)=\ln(x)
]
hasta
[
g(x)=-5\ln(x)
]

La idea es hacerlo en dos pasos:

1. **Dilatación vertical**: (\ln(x) \to 5\ln(x))
2. **Reflexión respecto al eje (x)**: (5\ln(x) \to -5\ln(x))

---

## 🔎 Visualización conceptual

Luego:

* Dilatación: (y=5\ln(x))
* Reflexión:

---

## 🐍 Código en **Matplotlib**

Este código muestra las tres funciones en la misma gráfica para ver claramente la transformación:

```python


````python
import numpy as np
import matplotlib.pyplot as plt

# Dominio (x > 0)
x = np.linspace(0.1, 5, 400)

# Funciones
y1 = np.log(x)        # ln(x)
y2 = 5 * np.log(x)    # dilatación
y3 = -5 * np.log(x)   # reflexión

# Gráfica
plt.figure(figsize=(8,6))

plt.plot(x, y1, label='ln(x)')
plt.plot(x, y2, label='5 ln(x)')
plt.plot(x, y3, label='-5 ln(x)')

plt.axhline(0)
plt.axvline(0)
plt.grid()

plt.legend()
plt.title("Secuencia de transformaciones de ln(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Dominio (x > 0)
x = np.linspace(0.1, 5, 400)

# Funciones
y1 = np.log(x)
y2 = 5 * np.log(x)
y3 = -5 * np.log(x)

# Crear matriz de gráficos
fig, axs = plt.subplots(1, 3, figsize=(15,5))

# Gráfico 1: ln(x)
axs[0].plot(x, y1)
axs[0].set_title("y = ln(x)")
axs[0].axhline(0)
axs[0].axvline(0)
axs[0].grid()

# Gráfico 2: 5ln(x)
axs[1].plot(x, y2)
axs[1].set_title("y = 5 ln(x) (Dilatación)")
axs[1].axhline(0)
axs[1].axvline(0)
axs[1].grid()

# Gráfico 3: -5ln(x)
axs[2].plot(x, y3)
axs[2].set_title("y = -5 ln(x) (Reflexión)")
axs[2].axhline(0)
axs[2].axvline(0)
axs[2].grid()

plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\transformaciones_ln.png", bbox_inches='tight')
plt.show()
````

## 🎯 **Secuencia de transformaciones (4 pasos)**

   [
   y = |x - \tfrac{1}{2}|
   ]

2. **Dilatación vertical**:
   [

   ]

4. **Traslación vertical hacia arriba**:
   [

## 🐍 **Código en Matplotlib (matriz 2×2)**

```python id="h3k2ld"


````python
import numpy as np
import matplotlib.pyplot as plt

# Dominio
x = np.linspace(-3, 4, 400)

# Transformaciones
y1 = np.abs(x)                        # |x|
y2 = np.abs(x - 0.5)                  # traslado horizontal
y3 = 6 * np.abs(x - 0.5)              # dilatación
y4 = -6 * np.abs(x - 0.5) + 5         # reflexión + traslado vertical

# Crear matriz 2x2
fig, axs = plt.subplots(2, 2, figsize=(10,8))

# Gráfico 1
axs[0,0].plot(x, y1)
axs[0,0].set_title("y = |x|")
axs[0,0].axhline(0)
axs[0,0].axvline(0)
axs[0,0].grid()

# Gráfico 2
axs[0,1].plot(x, y2)
axs[0,1].set_title("y = |x - 1/2|")
axs[0,1].axhline(0)
axs[0,1].axvline(0)
axs[0,1].grid()

# Gráfico 3
axs[1,0].plot(x, y3)
axs[1,0].set_title("y = 6|x - 1/2|")
axs[1,0].axhline(0)
axs[1,0].axvline(0)
axs[1,0].grid()

# Gráfico 4
axs[1,1].plot(x, y4)
axs[1,1].set_title("y = -6|x - 1/2| + 5")
axs[1,1].axhline(0)
axs[1,1].axvline(0)
axs[1,1].grid()

plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\transformaciones_valores_absolutos.png", bbox_inches='tight')
plt.show()

````
