# Prompt para diseño de función para toma de asistencia  

Dame una función de usuario de Python, para tomar asistencia en el espacio de asesoría, donde la función: consigne los datos tomados de asistencia en un archivo .xlsx. Te pido que la función le asigne un identificador corto a cada estudiante asistente a la asesoría, para que cada vez que asista de nuevo a la asesoría, no sea necesario preguntarle todos los datos sino solo ingresar el identificador o nombre corto para el estudiante y la dificultad o solicitud que lo trae al espacio de asesoría. Dame el diseño  de esta función, de tal manera que ella consigne, nombre del estudiante, correo o email, programa académico y dificultad o solicitud que lo trae al espacio a asesoría. 

```python
# Toma de asistencia rápida al espacio de asesoría  

from openpyxl import Workbook, load_workbook
import os  

def registrar_asistencia():
    """
    Función para registrar asistencia en un espacio de asesoría.
    Guarda nombre, correo, programa académico y dificultad/solicitud.
    El archivo se guarda y actualiza en la carpeta especificada.
    """

    # Ruta donde se guardará el archivo
    carpeta = r"C:\Users\marco\OneDrive - Universidad de Antioquia\Documentos\1_administracion\informes_jefe\renovacion_contrato_ocasional\asistencia_asesoria"
    archivo = os.path.join(carpeta, "asistencia.xlsx")

    # Crear carpeta si no existe
    os.makedirs(carpeta, exist_ok=True)

    # Verificar si existe el archivo, si no lo crea
    if os.path.exists(archivo):
        wb = load_workbook(archivo)
        ws = wb.active
    else:
        wb = Workbook()
        ws = wb.active
        ws.append(["ID", "Nombre", "Correo", "Programa", "Dificultad/Solicitud"])
    
    # Diccionario de estudiantes ya registrados (ID como clave)
    estudiantes = {}
    for row in ws.iter_rows(min_row=2, values_only=True):
        estudiantes[row[0]] = {"nombre": row[1], "correo": row[2], "programa": row[3]}
    
    # Pedir identificador o nombre corto
    identificador = input("Ingrese su identificador corto o nuevo ID: ").strip()
    
    if identificador in estudiantes:
        print(f"Bienvenido de nuevo {estudiantes[identificador]['nombre']}")
        dificultad = input("Ingrese la dificultad o solicitud que lo trae: ")
        ws.append([identificador, estudiantes[identificador]['nombre'], 
                   estudiantes[identificador]['correo'], 
                   estudiantes[identificador]['programa'], 
                   dificultad])
    else:
        print("Nuevo estudiante, por favor ingrese sus datos completos.")
        nombre = input("Nombre completo: ")
        correo = input("Correo electrónico: ")
        programa = input("Programa académico: ")
        dificultad = input("Ingrese la dificultad o solicitud que lo trae: ")
        
        estudiantes[identificador] = {"nombre": nombre, "correo": correo, "programa": programa}
        ws.append([identificador, nombre, correo, programa, dificultad])
    
    # Guardar cambios en la ruta indicada
    wb.save(archivo)
    print(f"Asistencia registrada en {archivo}")

```

```python
registrar_asistencia()
```

```python
def registrar_asistente():
    import csv
    import os
    from datetime import datetime

    # Solicitar información del asistente
    nombre = input("Ingrese el nombre del asistente: ")
    edad = input("Ingrese la edad del asistente: ")
    genero = input("Ingrese el género del asistente (M/F): ")
    correo = input("Ingrese el correo electrónico del asistente: ")
    curso = input("Ingrese el nombre del curso al que pertenece su inquietud: ")
    institucion = input("Ingrese la institución de donde viene: ")
    dificultad = input("Describa brevemente en qué consiste su dificultad: ")

    # Obtener la fecha actual en formato AAAA-MM-DD
    fecha = datetime.today().strftime('%Y-%m-%d')

    # Definir la ruta completa del archivo
    ruta_base = r"C:\Users\marco\Documentos\investigacion\investigacion_en_asesoria\2_secuencia_accion\3_Creacion_Comunidad_Aprendizaje"
    archivo = os.path.join(ruta_base, 'registro_asistentes.csv')
    encabezados = ['Nombre', 'Edad', 'Género', 'Correo', 'Curso', 'Institución', 'Dificultad', 'Fecha']

    # Verificar si la carpeta existe, si no, crearla
    if not os.path.exists(ruta_base):
        os.makedirs(ruta_base)

    # Verificar si el archivo ya existe
    archivo_existe = os.path.isfile(archivo)

    # Escribir los datos
    with open(archivo, 'a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f, delimiter=',')
        if not archivo_existe:
            writer.writerow(encabezados)
        writer.writerow([nombre, edad, genero, correo, curso, institucion, dificultad, fecha])

    print(f"Registro exitoso. Los datos se han guardado en: {archivo}")
```

```python
registrar_asistente()
```

## Título de la Asesoría: 



$$ |x-2| \le \frac{1}{5} $$

$$ \left|x - \frac{1}{3} \right| \leq \frac{3}{2} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalo solución
a = -7/6
b = 11/6

x = np.linspace(-4,4,400)

plt.figure(figsize=(8,2))

# Recta real
plt.axhline(0)

# Región solución
plt.plot([a,b],[0,0], linewidth=8)

# Puntos extremos
plt.scatter([a,b],[0,0])

plt.text(a,0.05,r"$-\frac{7}{6}$", ha='center')
plt.text(b,0.05,r"$\frac{11}{6}$", ha='center')

plt.yticks([])
plt.title(r"$\left|x-\frac{1}{3}\right|\leq \frac{3}{2}$")

plt.xlim(-4,4)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,4,400)
y = np.abs(x - 1/3)

plt.figure(figsize=(7,4))
plt.plot(x,y,label=r"$|x-\frac{1}{3}|$")
plt.axhline(3/2, linestyle="--", label=r"$\frac{3}{2}$")

plt.axvline(-7/6, linestyle=":")
plt.axvline(11/6, linestyle=":")
plt.text(-7/6,0.1,r"$-\frac{7}{6}$", ha='center')
plt.text(11/6,0.1,r"$\frac{11}{6}$ ", ha='center') 
plt.legend()
plt.title("Interpretación gráfica de la inecuación")

plt.show()
```

$$\left| \frac{2x}{11} + \frac{3}{17} \right| \le x-2 $$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,10,1000)

y1 = np.abs((2*x)/11 + 3/17)
y2 = x - 2

plt.figure(figsize=(7,5))

plt.plot(x,y1,label=r"$|2x/11 + 3/17|$")
plt.plot(x,y2,label=r"$x-2$")
plt.text(0,0.5,r"$|2x/11 + 3/17|$", ha='center')
# Punto de intersección
x_sol = 407/153
y_sol = x_sol - 2
plt.scatter([x_sol],[y_sol])

# región solución
mask = y1 <= y2
plt.fill_between(x,y1,y2,where=mask,alpha=0.3)

plt.axhline(0)

plt.legend()
plt.title("Solución gráfica de la inecuación")

plt.show()

print("Solución: x >=", 407/153)
```
