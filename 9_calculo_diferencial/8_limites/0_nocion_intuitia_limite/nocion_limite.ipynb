## Título de la Asesoría: 



$$f(x) = \frac{2x^{2} - x -1}{x-1}$$

Esta es Una forma muy didáctica de presentar la **idea intuitiva de límite** con Python combinando cálculo simbólico sencillo, una tabla con **Pandas** y una visualización con **Matplotlib** que muestre cómo la función se acerca a un valor aunque el punto no esté definido.

La función es

$$
f(x)=\frac{2x^{2}-x-1}{x-1}
$$



Primero observemos algo matemáticamente importante.

Factorizamos el numerador:

$$
2x^{2}-x-1=(2x+1)(x-1)
$$

Entonces, para (x\neq1):

$$
f(x)=2x+1
$$



Pero **en $x=1$ la función original no está definida**.

Sin embargo,

$$
\lim_{x\to1}f(x)=2(1)+1=3
$$

Esta es la idea clave para la clase:
**la función se acerca a 3 aunque en (x=1) no exista valor.**

---



# 1. Exploración numérica con Pandas

Primero hacemos una tabla de valores acercándonos a $x=1$.

```python
import numpy as np
import pandas as pd

def f(x):
    return (2*x**2 - x - 1)/(x - 1)

# valores alrededor de 1
x_vals = np.array([
    0.9, 0.99, 0.999,
    1.001, 1.01, 1.1
])

df = pd.DataFrame({
    "x": x_vals,
    "f(x)": f(x_vals)
})

print(df)
```


Salida esperada (aproximada):

| x     | f(x)  |
| ----- | ----- |
| 0.9   | 2.8   |
| 0.99  | 2.98  |
| 0.999 | 2.998 |
| 1.001 | 3.002 |
| 1.01  | 3.02  |
| 1.1   | 3.2   |



Interpretación para estudiantes:

* Cuando $x$ se acerca a **1 por la izquierda**, $f(x)$ se acerca a **3**
* Cuando $x$ se acerca a **1 por la derecha**, también se acerca a **3**

Entonces:

$$
\lim_{x\to1}f(x)=3
$$

---



# 2. Visualización gráfica de la idea de límite

Ahora mostramos la gráfica y el **hueco en $x=1$**.


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (2*x**2 - x - 1)/(x - 1)

x = np.linspace(-2, 4, 400)
y = f(x)

plt.figure(figsize=(8,5))

plt.plot(x, y, label="f(x)")

# punto del límite
plt.scatter(1, 3, marker='o', facecolors='none', s=120)

# línea vertical del punto problemático
plt.axvline(1)

plt.title("Idea intuitiva de límite")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid()
plt.legend()
plt.savefig('_.png')
plt.show()
```


Interpretación para clase:

* La gráfica es prácticamente la recta

$$
y = 2x+1
$$

* Pero hay un **hueco en (x=1)**.
* Aun así, **los valores se acercan a 3**.

---



# 3. Animación conceptual para estudiantes (muy potente en clase)

Podemos mostrar cómo los puntos se acercan al límite.


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (2*x**2 - x - 1)/(x - 1)

x = np.linspace(-2,4,400)
y = f(x)

plt.figure(figsize=(8,5))
plt.plot(x,y)

approach = [0.5,0.8,0.9,0.99,1.01,1.1,1.2]

for val in approach:
    plt.scatter(val,f(val))

plt.scatter(1,3, marker='o', facecolors='none', s=120)

plt.axvline(1)

plt.title("Acercamiento al límite")
plt.grid()
plt.savefig('_png')
plt.show()
```


# 4. Explicación pedagógica para tus estudiantes

Idea clave que puedes decir en clase:

1. Una función puede **no existir en un punto**.
2. Aun así, podemos preguntar:

> ¿A qué valor se acercan los resultados cuando nos aproximamos al punto?

3. Si ambos lados se acercan al mismo número, ese número es el **límite**.

En este ejemplo:

$$
\lim_{x\to1}\frac{2x^{2}-x-1}{x-1}=3
$$

aunque $f(1)$ **no está definida**.
