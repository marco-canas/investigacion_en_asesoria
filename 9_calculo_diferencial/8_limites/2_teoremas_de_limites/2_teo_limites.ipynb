# Teoremas de límites 


$$\lim_{x \to 4} 2x^{2} - 3x +  4$$

```python
# grafico de $$\lim_{x \to 4} 2x^{2} - 3x +  4$$
import numpy as np
import matplotlib.pyplot as plt

# función
def f(x):
    return 2*x**2 - 3*x + 4

# valores de x alrededor de 4
x = np.linspace(2,6,400)
y = f(x)

plt.figure(figsize=(7,5))

# gráfica de la función
plt.plot(x, y, label="f(x) = 2x² - 3x + 4")

# punto del límite
x0 = 4
y0 = f(x0)

plt.scatter([x0], [y0])
plt.axvline(x0)
plt.axhline(y0)

plt.title("Visualización del límite cuando x → 4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.xticks(np.arange(2, 6.5, 0.5))
plt.show()
```

$$\lim_{-2} \frac{x^{3} + 2x^{2} - 1}{5 - 3x}$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**3 + 2*x**2 - 1) / (5 - 3*x)

# valores alrededor de -2
x = np.linspace(-4,0,500)
y = f(x)

plt.figure(figsize=(7,5))

plt.plot(x, y, label="f(x) = (x³ + 2x² - 1)/(5 - 3x)", color='red')

# punto del límite
x0 = -2
y0 = f(x0)

plt.scatter([x0], [y0])
plt.axvline(x0)
plt.axhline(y0)

plt.title("Visualización del límite cuando x → -2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.yticks(np.arange(-2, 0.5, 0.1))
plt.show()
```

$$\lim_{x \to 1} \frac{x^{2}- 1}{x-1} $$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**2 - 1)/(x - 1)

# valores evitando x = 1
x1 = np.linspace(-2,0.99,400)
x2 = np.linspace(1.01,4,400)

plt.figure(figsize=(7,5))

plt.plot(x1, f(x1), label="f(x) = (x²-1)/(x-1)")
plt.plot(x2, f(x2))

# punto del límite (hueco)
plt.scatter([1],[2], facecolors='none', edgecolors='red', s=100)

plt.axvline(1)
plt.axhline(2)

plt.title("Visualización del límite cuando x → 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def f(x):
    return (x**2 - 1)/(x - 1)

def visualizar(epsilon=1.0):

    x_left = 1 - epsilon
    x_right = 1 + epsilon
    
    y_left = f(x_left)
    y_right = f(x_right)

    x = np.linspace(-2,4,400)

    plt.figure(figsize=(6,6))

    # gráfica
    plt.plot(x, x+1, label="y = x + 1")

    # hueco
    plt.scatter([1],[2], facecolors='white', edgecolors='black', s=80)

    # puntos sobre la curva
    plt.scatter([x_left,x_right],[y_left,y_right], s=80)

    # puntos sobre eje x
    plt.scatter([x_left,x_right],[0,0], s=80)

    # puntos sobre eje y
    plt.scatter([0,0],[y_left,y_right], s=80)

    # líneas guía
    plt.plot([x_left,x_left],[0,y_left], linestyle="dashed")
    plt.plot([x_right,x_right],[0,y_right], linestyle="dashed")

    plt.plot([0,x_left],[y_left,y_left], linestyle="dashed")
    plt.plot([0,x_right],[y_right,y_right], linestyle="dashed")

    # referencia límite
    plt.axvline(1)
    plt.axhline(2)

    plt.xlim(-2,4)
    plt.ylim(-1,5)

    plt.grid(True)
    plt.title("Visualización dinámica del límite")

    plt.show()


interact(
    visualizar,
    epsilon=widgets.FloatSlider(
        value=1,
        min=0.02,
        max=2,
        step=0.02,
        description="distancia"
    )
)
```
