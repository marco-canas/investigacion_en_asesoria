# Definición de las funciones trigonométricas de un ángulo agudo en un triángulo rectángulo

---

# 1. Situación geométrica inicial

Consideremos un **triángulo rectángulo**.

* Un ángulo es de (90^\circ)
* Tomamos un ángulo agudo (\theta)

Lados:

* **Hipotenusa**: lado opuesto al ángulo recto
* **Cateto opuesto** a (\theta)
* **Cateto adyacente** a $\theta$

---



# 2. Construcción del triángulo con Python

```python
import numpy as np
import matplotlib.pyplot as plt

# vértices del triángulo
A = (0,0)
B = (4,0)
C = (4,3)

x = [A[0], B[0], C[0], A[0]]
y = [A[1], B[1], C[1], A[1]]

plt.figure(figsize=(6,6))
plt.plot(x,y, marker='o')

plt.text(0,0,"A")
plt.text(4,0,"B")
plt.text(4,3,"C")

plt.text(2,0,"Cateto adyacente", ha='center')
plt.text(4,1.5,"Cateto opuesto", rotation=90, va='center')
plt.text(2.2,1.7,"Hipotenusa")

plt.title("Triángulo rectángulo")
plt.axis('equal')
plt.grid()
plt.show()
```

En esta figura:

$$
\theta = \angle A
$$

---

# 3. Identificación del ángulo

````python
import matplotlib.pyplot as plt
import numpy as np

A = np.array((0,0))
B = np.array((4,0))
C = np.array((4,3))

plt.figure(figsize=(6,6))

plt.plot([A[0],B[0]],[A[1],B[1]])
plt.plot([B[0],C[0]],[B[1],C[1]])
plt.plot([A[0],C[0]],[A[1],C[1]])

# arco del ángulo
theta = np.linspace(0, np.arctan(3/4), 100)
r = 0.8
plt.plot(r*np.cos(theta), r*np.sin(theta))

plt.text(0.6,0.2,r'$\theta$')

plt.axis('equal')
plt.grid()
plt.title("Ángulo agudo θ en el triángulo rectángulo")
plt.savefig(r"C:\Users\marco\Downloads\angulo_agudo.png")  # Guardar la figura como imagen
plt.show()
````

# 4. Definición de seno

El **seno** de un ángulo agudo es

$$
\sin(\theta)=
\frac{\text{cateto opuesto}}{\text{hipotenusa}}
$$

Ejemplo en el triángulo:

$$
\sin(\theta)=\frac{3}{5}
$$

porque

$$
hipotenusa=\sqrt{4^2+3^2}=5
$$

---

# 5. Visualización del seno

```python
import numpy as np
import matplotlib.pyplot as plt

adj = 4
opp = 3
hip = np.sqrt(adj**2 + opp**2)

A=(0,0)
B=(adj,0)
C=(adj,opp)

plt.figure(figsize=(6,6))

plt.plot([A[0],B[0],C[0],A[0]],
         [A[1],B[1],C[1],A[1]], marker='o')

plt.text(adj/2,-0.3,"adyacente = 4")
plt.text(adj+0.1,opp/2,"opuesto = 3")
plt.text(adj/2,opp/2,"hipotenusa = 5")

plt.title("Relación seno")

plt.axis('equal')
plt.grid()
plt.show()

print("sin(θ) =", opp/hip)
```

---

# 6. Definición de coseno

El **coseno** se define como

[
\cos(\theta)=
\frac{\text{cateto adyacente}}{\text{hipotenusa}}
]

En el triángulo

[
\cos(\theta)=\frac{4}{5}
]

---

# 7. Definición de tangente

La **tangente**

[
\tan(\theta)=
\frac{\text{cateto opuesto}}{\text{cateto adyacente}}
]

En el ejemplo

[
\tan(\theta)=\frac{3}{4}
]

---

# 8. Verificación con Python

```python
import numpy as np

adj = 4
opp = 3
hip = np.sqrt(adj**2 + opp**2)

print("sin =", opp/hip)
print("cos =", adj/hip)
print("tan =", opp/adj)
```

Resultado

```
sin = 0.6
cos = 0.8
tan = 0.75
```

---

# 9. Interpretación matemática importante

Las razones trigonométricas **no dependen del tamaño del triángulo**, solo del ángulo.

Si ampliamos el triángulo:

[
(4,3,5) \rightarrow (8,6,10)
]

las razones permanecen iguales.

---

# 10. Demostración visual

```python
import numpy as np
import matplotlib.pyplot as plt

escala = 2

adj = 4*escala
opp = 3*escala

A=(0,0)
B=(adj,0)
C=(adj,opp)

plt.figure(figsize=(6,6))
plt.plot([0,adj,adj,0],[0,0,opp,0], marker='o')

plt.axis('equal')
plt.grid()
plt.title("Triángulo semejante")

plt.show()

print("tan =", opp/adj)
```

Los estudiantes observan que:

[
\tan(\theta)=\frac{3}{4}
]

permanece igual.

---



# 11. Resumen conceptual

Para un ángulo agudo $\theta$:

$$
\sin(\theta)=\frac{\text{opuesto}}{\text{hipotenusa}}
$$

$$
\cos(\theta)=\frac{\text{adyacente}}{\text{hipotenusa}}
$$

$$
\tan(\theta)=\frac{\text{opuesto}}{\text{adyacente}}
$$

---

# Gráficos de las funciones trigonométricas. 



````python
import numpy as np
import matplotlib.pyplot as plt

# Dominio en grados
x_deg = np.linspace(0, 360, 2000)

# Convertir a radianes
x_rad = np.deg2rad(x_deg)

# Función tangente
y = np.tan(x_rad)

# Crear gráfica
plt.figure(figsize=(10,5))
plt.plot(x_deg, y, color = 'red')

# Límites para visualizar mejor
plt.ylim(-10, 10)

# Líneas verticales donde la tangente no está definida
for a in [90, 270]:
    plt.axvline(a, linestyle='--')

# Etiquetas
plt.xlabel("Ángulo (grados)")
plt.ylabel("tan(x)")
plt.xticks(np.arange(0, 361, 30))  # Marcas cada 30 grados
plt.hlines(0, 0, 360, color='blue', linewidth=0.5)  # Línea horizontal en y=0
plt.vlines(0, -10, 10, color='blue', linewidth=0.5)  # Línea vertical en x=0
plt.title("Gráfica de la función tangente de 0° a 360°")

# Cuadrícula
plt.grid(True)
plt.savefig(r"C:\Users\marco\Downloads\tangent_plot.png")  # Guardar la figura como imagen
plt.show()
````
