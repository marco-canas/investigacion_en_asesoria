# Funciones definidas a trozos  



$$f(x) = \begin{cases} 5 & x \le 2 \\ 2x -3  & x > 2 \end{cases} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Dominios
x1 = np.linspace(-5, 2, 200)   # x ≤ 2
x2 = np.linspace(2, 6, 200)    # x > 2

# Funciones
y1 = np.full_like(x1, 5)
y2 = 2*x2 - 3

plt.figure(figsize=(7,5))

# Primera parte
plt.plot(x1, y1, label=r"$f(x)=5,\; x\leq2$")

# Segunda parte
plt.plot(x2, y2, label=r"$f(x)=2x-3,\; x>2$")

# Punto cerrado
plt.scatter(2,5)

# Punto abierto
plt.scatter(2,2*2-3, facecolors='white', edgecolors='black')

plt.axhline(0)
plt.axvline(0)

plt.legend()
plt.title("Función definida por partes")
plt.grid(alpha=0.3) 
plt.yticks(np.arange(-5, 10, 1))
plt.xticks(np.arange(-5, 7, 1))
plt.savefig(r"C:\Users\marco\Downloads\funcion_partes.png", dpi=300)
plt.show()
```

$$f(x) = \begin{cases} x^{2} + 2x & x \le -1 \\ x & -1 < x \le 1 \\ -1 & x > 1 \end{cases} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Dominios
x1 = np.linspace(-5, -1, 200)
x2 = np.linspace(-1, 1, 200)
x3 = np.linspace(1, 5, 200)

# Funciones
y1 = x1**2 + 2*x1
y2 = x2
y3 = -1*np.ones_like(x3)

plt.figure(figsize=(8,5))

# Gráficas
plt.plot(x1, y1, label=r"$x^2+2x,\; x\leq -1$")
plt.plot(x2, y2, label=r"$x,\; -1<x\leq1$")
plt.plot(x3, y3, label=r"$-1,\; x>1$")

# Puntos especiales
plt.scatter(-1, (-1)**2 + 2*(-1))      # cerrado
plt.scatter(-1, -1, facecolors='white', edgecolors='black')  # abierto

plt.scatter(1, 1)   # cerrado
plt.scatter(1, -1, facecolors='white', edgecolors='black')  # abierto

# Ejes
plt.axhline(0)
plt.axvline(0)

plt.grid(alpha=0.3)
plt.legend()
plt.title("Función definida por partes")
plt.savefig(r"C:\Users\marco\Downloads\funcion_partes.png", dpi=300)
plt.show()
```

# Algoritmo o secuencia para graficar  una función definida a trozos  

1. trazo los ejes y los etiqueto. 
2. Metrizo los ejes. 
3. Trazo los puntos de división del dominio. 
4. Empieza a trazar desde la función más sencillas. 



```python
import numpy as np
import matplotlib.pyplot as plt

# Dominios válidos
x1 = np.linspace(-10, -3, 400)
x2 = np.linspace(3, 10, 400)

# Función
y1 = np.sqrt(x1**2 - 9)
y2 = np.sqrt(x2**2 - 9)

plt.figure(figsize=(7,5))

plt.plot(x1, y1, label=r"$\sqrt{x^2-9}$")
plt.plot(x2, y2)

# puntos extremos
plt.scatter([-3,3],[0,0])

plt.axhline(0)
plt.axvline(0)

plt.title(r"$f(x)=\sqrt{x^2-9}$")
plt.legend()
plt.grid(alpha=0.3)
plt.xticks(np.arange(-10, 11, 1))
plt.savefig(r"C:\Users\marco\Downloads\funcion_raiz.png", dpi=300)
plt.show()
```

# Las funciones elementales o básicas  

1. La función logaritmica $f(x) = \ln (x)$
2. La función exponencial $f(x) = e^{x}$. 
3. la función identidad $f(x) = x$
4. La función $f(x) = x^{2}$ 
5. la función elevar al cubo $f(x) = x^{3}$
6. La función raíz cuadrada $f(x) = \sqrt{x}$
7. La función valor valor absoluto $f(x) = |x|$. 


```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 3, figsize=(12,10))

# 1 Logaritmo
x = np.linspace(0.01,10,400)
axs[0,0].plot(x, np.log(x), color="blue")
axs[0,0].set_title(r"$\ln(x)$")
axs[0,0].axhline(0)
axs[0,0].axvline(0)

# 2 Exponencial
x = np.linspace(-3,3,400)
axs[0,1].plot(x, np.exp(x), color="magenta")
axs[0,1].set_title(r"$e^x$")
axs[0,1].axhline(0)
axs[0,1].axvline(0)

# 3 Identidad
axs[0,2].plot(x, x, color="brown")
axs[0,2].set_title(r"$x$")
axs[0,2].axhline(0)
axs[0,2].axvline(0)

# 4 Cuadrática
axs[1,0].plot(x, x**2, color="purple")
axs[1,0].set_title(r"$x^2$")
axs[1,0].axhline(0)
axs[1,0].axvline(0)

# 5 Cúbica
axs[1,1].plot(x, x**3, color="red")
axs[1,1].set_title(r"$x^3$")
axs[1,1].axhline(0)
axs[1,1].axvline(0)

# 6 Raíz
x = np.linspace(0,10,400)
axs[1,2].plot(x, np.sqrt(x), color="green")
axs[1,2].set_title(r"$\sqrt{x}$")
axs[1,2].axhline(0)
axs[1,2].axvline(0)

# 7 Valor absoluto
x = np.linspace(-5,5,400)
axs[2,0].plot(x, np.abs(x), color="orange")
axs[2,0].set_title(r"$|x|$")
axs[2,0].axhline(0)
axs[2,0].axvline(0)

# Quitar dos espacios vacíos
axs[2,1].axis("off")
axs[2,2].axis("off")

for ax in axs.flat:
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\funciones_dominio_rango.png", dpi=300)
plt.show()
```

La función

$$
f(x) = \ln(x)
$$

tiene estas propiedades importantes para la gráfica:

* **Dominio:** $x>0$
* **Pasa por:** $(1,0)$
* **Asíntota vertical:** $x=0$
* Crece lentamente.

# Script en Python con Matplotlib



```python
import numpy as np
import matplotlib.pyplot as plt

# Dominio (x > 0)
x = np.linspace(0.01, 10, 400)
y = np.log(x)

plt.figure(figsize=(7,5))

plt.plot(x, y, label=r"$\ln(x)$", color = "orange")

# Punto notable
plt.scatter([1],[0])

# Asíntota
plt.axvline(0, linestyle="--")

# Ejes
plt.axhline(0)
plt.axvline(0)

plt.title(r"Gráfica de $f(x)=\ln(x)$")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(alpha=0.3)
plt.legend()
plt.savefig(r"C:\Users\marco\Downloads\grafica_logaritmo.png", dpi=300)
plt.show()
```

# Interpretación geométrica para estudiantes

La función responde a la pregunta:

$$
e^{y}=x
$$

Por eso el logaritmo es la **inversa de la exponencial**.

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Dominios
x_exp = np.linspace(-3, 3, 400)
x_line = np.linspace(-10, 20, 400)
x_log = np.linspace(0.01, 15, 400)

# Funciones
y_exp = np.exp(x_exp)
y_line = x_line
y_log = np.log(x_log)

plt.figure(figsize=(8,6))

plt.plot(x_exp, y_exp, label=r"$e^x$")
plt.plot(x_line, y_line, label=r"$x$")
plt.plot(x_log, y_log, label=r"$\ln(x)$")

# Ejes
plt.axhline(0)
plt.axvline(0)

# Punto notable
plt.scatter([1],[1])

plt.xlim(-3,3)
plt.ylim(-3,10)

plt.grid(alpha=0.3)
plt.legend()
plt.axis('equal')
plt.title("Comparación: $e^x$, $x$, $\ln(x)$")
plt.savefig(r"C:\Users\marco\Downloads\comparacion_funciones.png", dpi=300)
plt.show()
```
