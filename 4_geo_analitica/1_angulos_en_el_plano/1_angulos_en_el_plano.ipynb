# Puntos terminales en la circunferencia unitaria

Suponga que $t$ es un número real. Marquemos una distancia $t$ a lo largo de la circunferencia unitaria, empezando en el punto $(1,0)$, y moviéndonos en dirección contraria al giro de las manecillas de un reloj si t es positiva y en el sentido de las manecillas si $t$ es negativa (Figura 2).   

En esta forma llegamos al punto P1x, y2 en la circunferencia. El punto P1x, y2 obtenido en esta forma se llama punto terminal determinado por el número real t.


Para lograr esto, necesitamos que el script calcule la posición del punto $P$ basándose en la definición del radián: en un círculo unitario, la longitud del arco $s$ es numéricamente igual al ángulo $\theta$ en radianes ($s = r \theta$).

Como quieres que $t$ controle la dirección, simplemente mapeamos $\theta = t$. 
* Si $t > 0$, el ángulo crece positivamente (antihorario).
* Si $t < 0$, el ángulo es negativo (horario).

Aquí tienes el código optimizado para **VS Code** (usando `%matplotlib tk` para evitar el error de importación previo):


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import matplotlib

matplotlib.use('TkAgg') 

def crear_animacion_arco():
    fig, ax = plt.subplots(figsize=(7, 7))
    plt.subplots_adjust(bottom=0.30)

    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='black', lw=0.8)
    ax.axvline(0, color='black', lw=0.8)
    ax.grid(True, linestyle=':', alpha=0.6)

    t_circle = np.linspace(0, 2*np.pi, 200)
    ax.plot(np.cos(t_circle), np.sin(t_circle), color='lightgray', ls='--')

    arco_rojo, = ax.plot([], [], color='red', lw=4, label='Arco longitud t')
    linea_radio, = ax.plot([], [], color='blue', lw=1, alpha=0.5)
    punto_p, = ax.plot([], [], 'ro', markersize=8)
    
    txt = ax.text(-1.4, 1.3, '', fontsize=10,
                  bbox=dict(facecolor='white', alpha=0.8))

    paso = np.pi / 12

    def actualizar(val):
        # 🔹 SOLO cuantizamos, NO tocamos el slider
        t = np.round(val / paso) * paso

        x = np.cos(t)
        y = np.sin(t)

        puntos_t = np.linspace(0, t, 100)
        arco_rojo.set_data(np.cos(puntos_t), np.sin(puntos_t))
        
        linea_radio.set_data([0, x], [0, y])
        punto_p.set_data([x], [y])
        
        txt.set_text(
            f't = {t/np.pi:.2f}π rad\n'
            f'P = ({x:.2f}, {y:.2f})'
        )
        
        fig.canvas.draw_idle()

    ax_slider = plt.axes([0.2, 0.15, 0.6, 0.03])

    slider = Slider(
        ax_slider, 't', -2*np.pi, 2*np.pi,
        valinit=0
    )

    # 🔹 Etiquetas en múltiplos de π
    ticks = np.arange(-2*np.pi, 2.1*np.pi, np.pi/2)
    labels = []

    for val in ticks:
        if np.isclose(val, 0):
            labels.append('0')
        else:
            frac = val / np.pi
            if np.isclose(frac, int(frac)):
                labels.append(f'{int(frac)}π')
            else:
                labels.append(f'{frac:.1f}π')

    ax_slider.set_xticks(ticks)
    ax_slider.set_xticklabels(labels, rotation=45)

    slider.on_changed(actualizar)

    actualizar(0)

    plt.title('Movimiento de P(x,y) según longitud de arco t')
    plt.legend(loc='lower left')

    return slider

if __name__ == "__main__":
    reg_slider = crear_animacion_arco()
    plt.show()
```

* **Sentido dinámico:** Al mover el deslizador hacia la izquierda (valores negativos), verás que el arco rojo se dibuja hacia abajo (sentido horario). Al moverlo a la derecha, sube (antihorario).
* **Resaltado del arco:** La variable `puntos_arco` genera una secuencia desde $0$ hasta $t$, lo que permite que Matplotlib trace la curva exacta sobre el perímetro del círculo.
* **Precisión matemática:** El punto $P(x, y)$ se calcula como $(\cos t, \sin t)$, cumpliendo con la identidad de que en el círculo unitario, la longitud del arco es igual al valor del ángulo en radianes.

## La Circunferencia Unitaria y la Longitud de Arco



La circunferencia de un círculo unitario (un círculo con radio $r = 1$) es $C = 2\pi(1) = 2\pi$. 



Si un punto $P$ inicia su recorrido en el punto $(1, 0)$ y se mueve en sentido contrario a las manecillas del reloj a lo largo de toda la vuelta del círculo unitario hasta regresar al punto $(1, 0)$, habrá viajado una distancia exacta de $2\pi$.

A partir de esta relación, podemos determinar las distancias recorridas y los puntos terminales para diferentes fracciones del círculo:

* **Media vuelta:** Para moverse la mitad del trayecto alrededor del círculo, el punto viaja una distancia de: 
    $$\frac{1}{2}(2\pi) = \pi$$
* **Un cuarto de vuelta:** Para moverse un cuarto de la distancia total alrededor del círculo, viaja una distancia de: 
    $$\frac{1}{4}(2\pi) = \frac{\pi}{2}$$
* **Tres cuartos de vuelta:** Para moverse tres cuartas partes de la distancia, recorre: 
    $$\frac{3}{4}(2\pi) = \frac{3\pi}{2}$$



### Puntos Terminales
¿Dónde termina el punto cuando viaja estas distancias a lo largo del círculo? Al observar el comportamiento en el plano cartesiano, vemos que:

1.  Cuando viaja una distancia $t = \pi$ iniciando en $(1, 0)$, su **punto terminal** es $(-1, 0)$.
2.  Cuando viaja una distancia $t = \frac{\pi}{2}$, su **punto terminal** es $(0, 1)$.
3.  Cuando viaja una distancia $t = \frac{3\pi}{2}$, su **punto terminal** es $(0, -1)$.





En general, para cualquier distancia $t$ recorrida sobre la circunferencia unitaria partiendo de $(1, 0)$, las coordenadas del punto terminal $P(x, y)$ están dadas por las funciones trigonométricas:
$$P(x, y) = (\cos t, \sin t)$$
