
# Demostremos la **inyectividad usando únicamente la definición**.

---



# 1. Definición de función inyectiva

Una función $f:A\to B$ es **inyectiva** si

$$
f(x_1)=f(x_2) \Rightarrow x_1=x_2
$$

---



# 2. Función dada

Consideremos

$$
f(x)=\frac{x}{\sqrt{x^2+1}}, \qquad x\in\mathbb{R}
$$



# 3. Supongamos igualdad de imágenes

Sean $x_1,x_2\in\mathbb{R}$ tales que

$$
f(x_1)=f(x_2)
$$

entonces

$$
\frac{x_1}{\sqrt{x_1^2+1}}=\frac{x_2}{\sqrt{x_2^2+1}}
$$

---



# 4. Multiplicación cruzada

$$
x_1\sqrt{x_2^2+1}=x_2\sqrt{x_1^2+1}
$$

---



# 5. Elevamos al cuadrado

(ambos lados son reales)

$$
x_1^2(x_2^2+1)=x_2^2(x_1^2+1)
$$

---



# 6. Expandimos

$$
x_1^2x_2^2+x_1^2=x_1^2x_2^2+x_2^2
$$

Restando $x_1^2x_2^2$ en ambos lados:

$$
x_1^2=x_2^2
$$

---



# 7. Consecuencia

$$
x_1=\pm x_2
$$

---



# 8. Eliminación del caso negativo

Si $x_1=-x_2$, entonces

$$
f(-x)=\frac{-x}{\sqrt{x^2+1}}=-f(x)
$$



Por lo tanto

$$
f(x_1)=f(-x_1)=-f(x_1)
$$

lo cual solo ocurre si

$$
f(x_1)=0
$$

y esto sucede únicamente cuando

$$
x_1=0
$$

En ese caso también $x_2=0$.

---



# 9. Conclusión

La única posibilidad compatible con

$$
f(x_1)=f(x_2)
$$

es

$$
x_1=x_2
$$



Por lo tanto,

$$
f(x)=\frac{x}{\sqrt{x^2+1}}
$$

es **inyectiva en $\mathbb{R}$**.

---



✅ Si quieres, también puedo mostrarte **una demostración aún más elegante (muy usada en cursos de Análisis)** basada en esta identidad:

$$
f(x_1)-f(x_2)=\frac{x_1-x_2}{\sqrt{x_1^2+1}\sqrt{x_2^2+1}}\left(\frac{\sqrt{x_1^2+1}+\sqrt{x_2^2+1}}{,}\right)
$$

que permite ver **directamente que el signo depende de $x_1-x_2$**, y por tanto la función es inyectiva.


```python
# gráfico de f(x) = x/(\sqrt{x^{2} + 1})
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 1000)
y = x / np.sqrt(x**2 + 1)
plt.plot(x, y)
plt.title('Gráfico de f(x) = x/(\sqrt{x^{2} + 1})')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.savefig(r'C:\Users\marco\Downloads\grafico.png')  # Salva o gráfico como um arquivo PNG
plt.show()
```
