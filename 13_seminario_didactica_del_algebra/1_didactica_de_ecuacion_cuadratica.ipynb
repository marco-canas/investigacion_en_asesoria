# Enfoque de enseñanza de la función cuadrática sin seguir los DBA, es decir desde un punto de vista abstracto 




Un **enfoque abstracto** para la enseñanza de la **función cuadrática** (sin seguir los lineamientos curriculares como los DBA del MEN) parte de la **estructura matemática del objeto**, su **naturaleza algebraica**, y sus **propiedades geométricas** dentro del estudio de las funciones polinómicas. Este enfoque es más cercano al que se usa en **matemáticas puras** o en cursos iniciales de **álgebra y análisis**.

---

# Enfoque abstracto para la enseñanza de la función cuadrática

## 1. La función como objeto matemático

El punto de partida es la **definición de función** como una aplicación entre conjuntos:

[
f: \mathbb{R} \rightarrow \mathbb{R}
]

donde a cada número real (x) le corresponde **un único número real** (f(x)).

Dentro del conjunto de funciones reales se estudia la familia de las **funciones polinómicas**.

---

## 2. Definición de función cuadrática

Una **función cuadrática** es una función polinómica de grado dos definida por

[
f(x)=ax^2+bx+c
]

donde

* (a,b,c \in \mathbb{R})
* (a \neq 0)

genui{"math_block_widget_common_keywords": {"content": "f(x)=ax^2+bx+c"}}

Esta expresión define un **subconjunto particular del espacio vectorial de los polinomios reales**.

Observaciones importantes:

* Si (a=0) la función deja de ser cuadrática y se vuelve lineal.
* El **grado del polinomio** determina propiedades globales de la función.

---

# 3. Interpretación estructural del polinomio

La función cuadrática puede estudiarse desde varias **representaciones algebraicas equivalentes**.

### Forma general

[
f(x)=ax^2+bx+c
]

Esta forma es útil para:

* identificar coeficientes
* estudiar raíces
* estudiar discriminante

---

### Forma factorizada

Si existen raíces reales (r_1) y (r_2):

[
f(x)=a(x-r_1)(x-r_2)
]

Aquí se destaca la **estructura multiplicativa** del polinomio.

---

### Forma canónica (forma de vértice)

[
f(x)=a(x-h)^2+k
]

Esta forma permite identificar directamente:

* vértice $(h,k)$
* eje de simetría
* desplazamientos

---



# 4. Interpretación geométrica

La gráfica de toda función cuadrática es una **parábola**.





Propiedades fundamentales:

1. **Simetría**

Existe una recta vertical llamada **eje de simetría**

$$
x=-\frac{b}{2a}
$$

2. **Vértice**

El punto donde la función alcanza un máximo o mínimo.

3. **Concavidad**

* $a>0$ → parábola abre hacia arriba
* $a<0$ → parábola abre hacia abajo

---



# 5. Estudio algebraico de las raíces

Las raíces se obtienen resolviendo

$$
ax^2+bx+c=0
$$


El término

$$
\Delta=b^2-4ac
$$

determina la naturaleza de las soluciones.

| Discriminante | Soluciones        |
| ------------- | ----------------- |
| $\Delta>0$    | dos raíces reales |
| $\Delta=0$    | raíz doble        |
| $\Delta<0$    | raíces complejas  |

---



# 6. Propiedades analíticas

Desde el **análisis matemático** se estudian propiedades como:

### Dominio

$$
Dom(f)=\mathbb{R}
$$

### Imagen

Depende del vértice:

* Si $a>0$

$$
Im(f)=[k,\infty)
$$

* Si (a<0)

$$
Im(f)=(-\infty,k]
$$

---



# 7. Relación con cálculo diferencial

La función cuadrática permite introducir ideas de cálculo.

$$
f'(x)=2ax+b
$$

El vértice aparece cuando

$$
f'(x)=0
$$

lo que conduce nuevamente a

$$
x=-\frac{b}{2a}
$$

---



# 8. Generalización dentro de los polinomios

La función cuadrática se ubica en la jerarquía

$$
\text{constantes} \subset \text{lineales} \subset \text{cuadráticas} \subset \text{polinomiales}
$$

Esto permite estudiar:

* crecimiento de grado
* comportamiento en el infinito
* número máximo de raíces

---



# 9. Actividades de pensamiento matemático (enfoque abstracto)

### 1

Demostrar que toda función cuadrática tiene **un único vértice**.

### 2

Probar que la gráfica es **simétrica respecto a una recta vertical**.

### 3

Demostrar que el número máximo de raíces reales es **dos**.

### 4

Transformar la forma general a la forma de vértice mediante **completar cuadrados**.

### 5

Mostrar que el conjunto de funciones cuadráticas forma un **espacio vectorial de dimensión 3**.

---



- **Idea central de este enfoque**

La función cuadrática no se presenta como una “fórmula para resolver problemas”, sino como:

* un **objeto algebraico**
* un **objeto geométrico**
* un **objeto analítico**

que puede estudiarse desde distintas representaciones.
