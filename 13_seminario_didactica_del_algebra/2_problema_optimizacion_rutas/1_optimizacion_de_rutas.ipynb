## Título de la Asesoría: 



El problema presentado en la imagen corresponde a una **interpretación aplicada del** Traveling Salesman Problem (Problema del Viajante), uno de los problemas clásicos de la **optimización combinatoria**. A continuación te doy una **interpretación conceptual, logística y matemática** del reto.

---

## 1. Interpretación del contexto real (logístico)

La empresa **“Entrega Rápida”** tiene un **almacén central (A)** desde donde salen los camiones para realizar entregas en cinco puntos:

[
B, C, D, E, F
]

El objetivo es encontrar una **ruta óptima** que:

1. Empiece en el almacén (A).
2. Pase **exactamente una vez por cada punto de entrega**.
3. Regrese nuevamente al almacén (A).

El problema consiste en **decidir el orden de visitas** a los puntos de entrega para **minimizar el costo total de la ruta**.

En términos logísticos, esto significa:

* reducir **distancia recorrida**,
* reducir **tiempo total de operación**,
* disminuir **consumo de combustible** y **costos operativos**.

---

## 2. Interpretación de las variables del problema

En la imagen aparecen dos tipos de costos:

### 1️⃣ Distancias entre puntos

Las distancias entre nodos del sistema logístico están aproximadamente entre:

[
4 \text{ km } \le d_{ij} \le 12 \text{ km}
]

donde:

[
d_{ij} = \text{distancia entre los puntos } i \text{ y } j
]

Estas distancias representan **el costo de transporte**.

---

### 2️⃣ Tiempo de entrega en cada parada

Cada punto de entrega requiere un **tiempo de descarga** entre:

[
10 \text{ min} \le t_i \le 18 \text{ min}
]

Esto representa **el costo operativo asociado al servicio en cada cliente**.

---

## 3. Interpretación del modelo matemático

El costo total de una ruta (R) se modela como:

[
C(R) = \sum \text{Distancias} + \sum \text{Tiempos de entrega}
]

Más formalmente:

[
C(R) =
\sum_{(i,j)\in R} d_{ij}
+
\sum_{i\in R} t_i
]

donde:

* (d_{ij}) = distancia entre nodos consecutivos
* (t_i) = tiempo de servicio en el nodo (i)

El objetivo del problema es:

[
\min C(R)
]

---

## 4. Interpretación combinatoria

Si hay **5 puntos de entrega**, el número de rutas posibles crece rápidamente.

La cantidad de rutas distintas (ignorando simetrías) es aproximadamente:

[
\frac{(n-1)!}{2}
]

Para (n = 6) nodos (incluyendo el almacén):

[
\frac{(6-1)!}{2} = \frac{120}{2} = 60
]

Esto muestra una característica importante:

### crecimiento factorial

[
n! = 1, 2, 6, 24, 120, 720, \dots
]

Por eso el problema se vuelve **computacionalmente complejo** cuando aumenta el número de destinos.

---

## 5. Interpretación en términos de teoría de grafos

El sistema puede representarse como un **grafo ponderado**:

* **Vértices:**
  (V = {A,B,C,D,E,F})

* **Aristas:**
  caminos entre puntos.

* **Pesos:**
  distancia + tiempo.

El problema consiste en encontrar un **ciclo Hamiltoniano mínimo**.

---

## 6. Interpretación computacional

Este problema aparece en muchos sistemas reales:

* logística de entregas
* rutas de drones
* planificación de transporte
* circuitos de manufactura
* rutas de recolección de basura

Por eso se estudia en áreas como:

* investigación de operaciones
* inteligencia artificial

# 1️⃣ Modelar el problema como un grafo con `networkx`

Primero representamos el sistema logístico como un **grafo ponderado**:

* **nodos** → almacén y puntos de entrega
* **aristas** → rutas posibles
* **peso** → distancia + tiempo de entrega

### Script



```python
import networkx as nx
import pandas as pd

# Crear grafo
G = nx.Graph()

# Nodos (almacén y puntos de entrega)
nodos = ["A","B","C","D","E","F"]
G.add_nodes_from(nodos)

# Distancias entre puntos (km)
distancias = [
("A","B",6),
("A","C",8),
("A","D",7),
("B","C",5),
("B","E",9),
("C","D",4),
("C","F",7),
("D","E",6),
("E","F",5)
]

# Tiempos de descarga (min)
tiempos = {
"A":0,
"B":12,
"C":15,
"D":10,
"E":18,
"F":14
}

# Agregar aristas con pesos
for u,v,d in distancias:
    costo = d + tiempos[v]
    G.add_edge(u,v,weight=costo,dist=d)

# Ver aristas
df = pd.DataFrame([
(u,v,data["weight"]) for u,v,data in G.edges(data=True)
],columns=["Origen","Destino","Costo"])

print(df)
```

Interpretación para estudiantes:

$$
\text{Costo} = \text{Distancia} + \text{Tiempo de descarga}
$$

---



# 2. Resolver el problema (ruta óptima)

Usamos el algoritmo aproximado del TSP incluido en `networkx`.

### Script


```python
from networkx.algorithms.approximation import traveling_salesman_problem

ruta = traveling_salesman_problem(G, weight='weight')

print("Ruta óptima aproximada:")
print(ruta)
```


También podemos calcular el costo total.



```python
costo_total = 0

for i in range(len(ruta)-1):
    costo_total += G[ruta[i]][ruta[i+1]]["weight"]

print("Costo total:", costo_total)
```


Esto permite discutir en clase:

* **algoritmos exactos**
* **algoritmos aproximados**
* complejidad computacional

# 3. Visualizar la ruta con `plotly`

Esto es muy potente para enseñanza porque muestra **la red logística y la ruta encontrada**.

### Script


```python
import plotly.graph_objects as go

pos = nx.spring_layout(G, seed=10)

edge_x = []
edge_y = []

for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    line=dict(width=1),
    hoverinfo='none',
    mode='lines'
)

node_x = []
node_y = []

for node in G.nodes():
    x, y = pos[node]
    node_x.append(x)
    node_y.append(y)

node_trace = go.Scatter(
    x=node_x,
    y=node_y,
    mode='markers+text',
    text=list(G.nodes()),
    textposition="top center",
    marker=dict(size=20)
)

fig = go.Figure(data=[edge_trace, node_trace])

fig.update_layout(
    title="Red logística de entregas",
    showlegend=False
)

fig.show()
```


# 4. Visualizar la **ruta óptima resaltada**

Podemos pintar la ruta en rojo.



```python
ruta_x = []
ruta_y = []

for i in range(len(ruta)-1):
    x0,y0 = pos[ruta[i]]
    x1,y1 = pos[ruta[i+1]]
    ruta_x += [x0,x1,None]
    ruta_y += [y0,y1,None]

ruta_trace = go.Scatter(
    x=ruta_x,
    y=ruta_y,
    mode="lines",
    line=dict(width=4,color="red"),
    name="Ruta óptima"
)

fig = go.Figure(data=[edge_trace,node_trace,ruta_trace])
fig.show()
```
