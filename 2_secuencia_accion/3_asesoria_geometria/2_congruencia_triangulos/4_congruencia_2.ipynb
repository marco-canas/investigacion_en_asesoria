# Demostrar que en un triángulo, a lados congruentes se oponen ángulos congruentes. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas del triángulo isósceles
A = (0, 2)
B = (-1, 0)
C = (1, 0)

# Coordenada del punto H (proyección de A sobre BC)
H = (0, 0)

# Crear la figura y los ejes
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Primer gráfico: Triángulo isósceles
axes[0].plot([A[0], B[0]], [A[1], B[1]], 'g')
axes[0].plot([A[0], C[0]], [A[1], C[1]], 'g')
axes[0].plot([B[0], C[0]], [B[1], C[1]], 'b')

axes[0].scatter(*zip(A, B, C), color='black')
axes[0].text(A[0], A[1]+0.1, 'A', fontsize=12, ha='right')
axes[0].text(B[0]-0.1, B[1], 'B', fontsize=12, ha='right')
axes[0].text(C[0]+0.1, C[1], 'C', fontsize=12, ha='left')
axes[0].set_title("Triángulo isósceles $\Delta ABC$")
axes[0].axis('equal')

# Segundo gráfico: Triángulo con altura
axes[1].plot([A[0], B[0]], [A[1], B[1]], 'b')
axes[1].plot([A[0], C[0]], [A[1], C[1]], 'b')
axes[1].plot([B[0], C[0]], [B[1], C[1]], 'b')
axes[1].plot([A[0], H[0]], [A[1], H[1]], 'r--', label="Altura $\overline{AH}$")
axes[1].scatter(*zip(A, B, C, H), color='black')
axes[1].set_title("Triángulo con altura $\overline{AH}$")
axes[1].axis('equal')

# Tercer gráfico: Triángulos $\Delta AHB$ y $\Delta AHC$
axes[2].fill([A[0], H[0], B[0]], [A[1], H[1], B[1]], color='orange', alpha=0.5, label="$\Delta AHB$")
axes[2].fill([A[0], H[0], C[0]], [A[1], H[1], C[1]], color='green', alpha=0.5, label="$\Delta AHC$")
axes[2].plot([B[0], C[0]], [B[1], C[1]], 'b')
axes[2].scatter(*zip(A, B, C, H), color='black')
axes[2].set_title("Triángulos $\Delta AHB$ y $\\Delta AHC$")
axes[2].axis('equal')

# Ajustar diseño
plt.tight_layout()

plt.savefig(r"C:/Users/marco/Downloads/triangulo_isosceles.png", dpi=300)
plt.show()
```

# Demostración con la estructura Afirmación - Razón 

|Afirmación|Razón|
|-|-|
|1. $\Delta ABC$, $AB = AC$|Hipótesis|
|2. Trazo la altura $\bar{AH}$|Construcción Auxiliar|
|3. $\bar{AH} \cong \bar{AH}$|Reflexividad de $\cong$|
|4. $\Delta AHB \cong \Delta AHC$ | de 2. de 1. y de 3. Y Criterio HC|
|5. $\hat{B} \cong \hat{C}$|de 4.: Ángulos correspondientes en triangulos congruentes|

Entonces, efectivamente, cada vez que tengamos un triángulo, en el que dos de sus lados sean congruentes, entonces, podremos afirmar que los ángulos opuestos a esos lados también son congruentes.
