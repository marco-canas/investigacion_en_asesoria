# Domos Geodésicos: Construcción y Visualización con Matplotlib

Los domos geodésicos son estructuras esféricas o semiesféricas compuestas por una red de triángulos que forman una superficie aproximada a una esfera. Fueron popularizados por Buckminster Fuller en el siglo XX y son conocidos por su eficiencia estructural y estética.

## Conceptos básicos

Un domo geodésico se define por su frecuencia (v), que indica cuántas divisiones tiene cada arista del icosaedro base. Las frecuencias comunes son 1v, 2v, 3v, etc.

## Construcción de un domo geodésico

Para construir un domo geodésico necesitamos:

1. Partir de un icosaedro (20 caras triangulares)
2. Subdividir cada triángulo según la frecuencia elegida
3. Proyectar los vértices sobre una esfera
4. Calcular las coordenadas de los vértices

## Implementación en Python con Matplotlib


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def crear_domo_geodesico(frecuencia=2, radio=1, hemisferio=True):
    # Ratio áureo
    phi = (1 + np.sqrt(5)) / 2
    
    # Vértices de un icosaedro
    vertices = [
        (-1, phi, 0), (1, phi, 0), (-1, -phi, 0), (1, -phi, 0),
        (0, -1, phi), (0, 1, phi), (0, -1, -phi), (0, 1, -phi),
        (phi, 0, -1), (phi, 0, 1), (-phi, 0, -1), (-phi, 0, 1)
    ]
    
    # Caras del icosaedro (20 triángulos)
    caras = [
        [0, 11, 5], [0, 5, 1], [0, 1, 7], [0, 7, 10], [0, 10, 11],
        [1, 5, 9], [5, 11, 4], [11, 10, 2], [10, 7, 6], [7, 1, 8],
        [3, 9, 4], [3, 4, 2], [3, 2, 6], [3, 6, 8], [3, 8, 9],
        [4, 9, 5], [2, 4, 11], [6, 2, 10], [8, 6, 7], [9, 8, 1]
    ]
    
    # Convertir a arrays numpy
    vertices = [np.array(v) for v in vertices]
    vertices = [v/np.linalg.norm(v) for v in vertices]  # Normalizar
    
    # Subdivisión de los triángulos
    def subdividir_triangulo(v1, v2, v3, freq):
        if freq == 0:
            return [[v1, v2, v3]]
        
        # Puntos intermedios
        v12 = (v1 + v2) / 2
        v23 = (v2 + v3) / 2
        v31 = (v3 + v1) / 2
        
        # Normalizar para proyectar sobre la esfera
        v12 = v12 / np.linalg.norm(v12)
        v23 = v23 / np.linalg.norm(v23)
        v31 = v31 / np.linalg.norm(v31)
        
        # Subdividir recursivamente
        triangulos = []
        triangulos += subdividir_triangulo(v1, v12, v31, freq-1)
        triangulos += subdividir_triangulo(v2, v23, v12, freq-1)
        triangulos += subdividir_triangulo(v3, v31, v23, freq-1)
        triangulos += subdividir_triangulo(v12, v23, v31, freq-1)
        
        return triangulos
    
    # Subdividir todas las caras
    triangulos = []
    for cara in caras:
        v1, v2, v3 = vertices[cara[0]], vertices[cara[1]], vertices[cara[2]]
        triangulos += subdividir_triangulo(v1, v2, v3, frecuencia)
    
    # Filtrar para hemisferio si es necesario
    if hemisferio:
        triangulos = [t for t in triangulos if all(v[2] >= 0 for v in t)]
    
    # Escalar por el radio y convertir a arrays numpy
    triangulos = [np.array([v*radio for v in t]) for t in triangulos]
    
    return triangulos

def visualizar_domo(triangulos):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Crear una colección de polígonos 3D
    coleccion = Poly3DCollection(triangulos, alpha=0.8, linewidths=1, edgecolor='k')
    coleccion.set_facecolor('cyan')
    
    ax.add_collection3d(coleccion)
    
    # Ajustar los límites
    max_range = np.max([np.max(np.abs(t)) for t in triangulos])
    ax.set_xlim([-max_range, max_range])
    ax.set_ylim([-max_range, max_range])
    ax.set_zlim([0, max_range])
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    ax.set_title(f'Domo Geodésico {len(triangulos)} triángulos)')
    plt.tight_layout()
    plt.show()

# Crear y visualizar un domo de frecuencia 2
triangulos = crear_domo_geodesico(frecuencia=2, radio=5)
visualizar_domo(triangulos)
````


## Variaciones y mejoras

1. **Diferentes frecuencias**: Cambia el parámetro `frecuencia` para obtener domos más o menos detallados.
2. **Domo completo vs hemisferio**: El parámetro `hemisferio` controla si se genera media esfera o una completa.
3. **Colores**: Puedes modificar el color de las caras cambiando `set_facecolor`.
4. **Animaciones**: Puedes crear animaciones rotando el domo con `ax.view_init`.

## Consideraciones estructurales

- Los domos de mayor frecuencia tienen más triángulos y se aproximan mejor a una esfera.
- En construcción real, los domos geodésicos usan diferentes longitudes de barras que pueden calcularse a partir de estas coordenadas.
- La estructura es extremadamente fuerte para su peso debido a la distribución uniforme de tensiones.

Este código te proporciona una base para explorar y visualizar domos geodésicos de diferentes configuraciones.

La relación entre el diámetro de una cúpula geodésica y la longitud de los lados de sus triángulos depende de la **frecuencia (v)** de la estructura y del **tipo de subdivisión** utilizada (Clase I, II o III). Aquí te explico cómo se relacionan:

---

## 1. **Frecuencia (v) y número de triángulos**
La frecuencia define cuántas veces se subdivide cada arista del icosaedro base. Por ejemplo:
- **1v**: 20 triángulos (icosaedro original).
- **2v**: 80 triángulos (cada arista se divide en 2).
- **3v**: 180 triángulos (cada arista se divide en 3).

A mayor frecuencia, más triángulos y más pequeñas son sus aristas para un mismo diámetro.

---

## 2. **Longitud de las aristas en función del diámetro**
En una cúpula geodésica perfectamente esférica:
- Todos los vértices están sobre una esfera de radio **R = Diámetro / 2**.
- Las aristas **no son todas iguales** (excepto en el caso 1v, que es un icosaedro regular).
- Existen **2 o 3 longitudes distintas** de aristas (dependiendo de la subdivisión).

### **Fórmula general para la longitud de una arista (l)**
Para una frecuencia **v**, la longitud de las aristas se puede aproximar por:
\[
l \approx \frac{\pi \cdot D}{2 \cdot v}
\]
Donde:
- \(D\) = Diámetro de la cúpula.
- \(v\) = Frecuencia.

Pero esto es solo una aproximación. La longitud exacta depende de la **posición del triángulo** en la estructura.

---

## 3. **Ejemplo práctico (2v)**
Para una cúpula de **frecuencia 2** y diámetro **6 metros**:
- **Radio (R) = 3 metros**.
- **Longitud de aristas**:
  - Las aristas más largas (las que conectan los pentágonos) ≈ **3.82 m**.
  - Las aristas más cortas (las internas de los hexágonos) ≈ **3.30 m**.

Esto se debe a que en una subdivisión geodésica, no todos los triángulos son iguales.

---

## 4. **Cómo calcularlo exactamente**
Para obtener las longitudes exactas, debes:
1. **Generar las coordenadas de los vértices** (como en el código de Python que compartí).
2. **Calcular la distancia euclidiana entre puntos adyacentes**:
   \[
   l = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}
   \]

Ejemplo en Python:
```python
import numpy as np

# Supongamos que tenemos dos vértices de un triángulo
v1 = np.array([1.0, 0.0, 0.0])
v2 = np.array([0.0, 1.0, 0.0])

# Longitud de la arista entre ellos
longitud = np.linalg.norm(v2 - v1)
print(f"Longitud de la arista: {longitud:.2f} metros")
```

---

## 5. **Relación en construcción real**
En domos reales (como los de bambú o acero), se usan **barras de diferentes longitudes**, agrupadas en:
- **Estratos (A, B, C)**: Según su posición en la cúpula.
- **Fórmulas empíricas**: Para calcular las longitudes sin modelado 3D.

Por ejemplo, en un domo **3v**, hay **3 longitudes distintas** de barras.

---

### **Conclusión**
- La longitud de las aristas **disminuye al aumentar la frecuencia**.
- No todas las aristas son iguales (excepto en 1v).
- Para un diámetro fijo, puedes calcular las longitudes exactas con las coordenadas de los vértices.
