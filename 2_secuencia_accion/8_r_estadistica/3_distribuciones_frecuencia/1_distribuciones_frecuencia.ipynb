## ✅ Opción 2: Generar un diagrama textual en consola (sin librerías externas)


````python
from collections import defaultdict

# Datos
likes = [145, 287, 356, 198, 423, 
         267, 389, 156, 234, 445, 
         312, 167, 398, 278, 189, 
         334, 256, 412, 523, 89]

# Crear estructura de tallo y hojas
stems = defaultdict(list)
for number in sorted(likes):
    stem, leaf = divmod(number, 100)
    stems[stem].append(leaf)

# Mostrar diagrama en consola
print("Diagrama de Tallo y Hojas")
for stem, leaves in sorted(stems.items()):
    print(f"{stem} | {' '.join(str(leaf) for leaf in leaves)}")

````

Perfecto, Marco 🙌. Te dejo un **script en Python** que calcula todas las medidas estadísticas que mencionas (**media, mediana, moda y rango**) a partir de los datos de likes.

```python
import statistics as stats

# Datos de la actividad
likes = [145, 287, 356, 198, 423, 
         267, 389, 156, 234, 445, 
         312, 167, 398, 278, 189, 
         334, 256, 412, 523, 89]

print("=== Medidas Estadísticas ===")

# Media
media = stats.mean(likes)
print(f"Media: {media:.2f}")

# Mediana
mediana = stats.median(likes)
print(f"Mediana: {mediana}")

# Moda
try:
    moda = stats.mode(likes)
    print(f"Moda: {moda}")
except stats.StatisticsError:
    print("No hay una moda única (los datos son multimodales)")

# Rango (máximo - mínimo)
rango = max(likes) - min(likes)
print(f"Rango: {rango}")
```

* **Media** ≈ 292.05
* **Mediana** = 282.5
* **Moda** → no hay moda única (todos los valores son distintos)
* **Rango** = 434 (523 – 89)

Para hacer un histograma bien construido necesitamos calcular el **número adecuado de intervalos de clase**. Una fórmula muy usada es la de **Sturges**:



$$
k = 1 + 3.322 \cdot \log_{10}(n)
$$

donde $n$ es el número de datos.

Aquí te dejo el **código corregido y mejorado** para construir el histograma de la variable `"Likes"`:


````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

# Datos
likes = [145, 287, 356, 198, 423, 
         267, 389, 156, 234, 445, 
         312, 167, 398, 278, 189, 
         334, 256, 412, 523, 89]

# Crear DataFrame
df = pd.DataFrame({"Likes": likes})

# ============================
# 1. Calcular número de clases con fórmula de Sturges
# ============================
n = len(df["Likes"])
k = int(1 + 3.322 * math.log10(n))  # Número de intervalos de clase
print(f"Número de clases según Sturges: {k}")

# ============================
# 2. Histograma
# ============================
plt.figure(figsize=(8,6))
bins = 
plt.xticks(np.arange(min(df['Likes']), ))
plt.hist(df["Likes"], bins=k, color="purple", edgecolor="black", alpha=0.7)

plt.title("Histograma de Likes (con regla de Sturges)", fontsize=14, fontweight="bold")
plt.xlabel("Intervalos de Likes")
plt.ylabel("Frecuencia")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

````


---

📌 Con tus datos ($n=20$):

$$
k = 1 + 3.322 \cdot \log_{10}(20) \approx 6
$$

El histograma se generará con **6 clases** bien definidas, mostrando cómo se distribuyen los likes de los estudiantes.

---

````python

# =========================
# 2. Diagrama de caja y bigotes
# =========================
plt.figure(figsize=(6,6))
plt.boxplot(df["Likes"], patch_artist=True,
            boxprops=dict(facecolor="orange", color="purple"),
            whiskerprops=dict(color="purple"),
            medianprops=dict(color="red", linewidth=2),
            capprops=dict(color="purple"))

plt.title("Diagrama de Caja y Bigotes de Likes", fontsize=14, fontweight="bold")
plt.ylabel("Likes")
plt.show()

````
