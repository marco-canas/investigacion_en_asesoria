La **medida de desempeño (R^2)** (llamada **coeficiente de determinación**) es un indicador fundamental en **modelos de regresión**, y aunque se suele estudiar dentro de la **estadística inferencial** y el **análisis de regresión**, tiene un fuerte vínculo con la **estadística descriptiva**, porque está construido a partir de **varianzas y sumas de cuadrados**.

---



## 🔹 ¿Qué es (R^2)?

Es una medida que indica **qué proporción de la variabilidad total de una variable dependiente (Y) es explicada por el modelo de regresión con las variables independientes (X)**.

[
R^2 = \frac{SS_{\text{reg}}}{SS_{\text{tot}}} = 1 - \frac{SS_{\text{res}}}{SS_{\text{tot}}}
]

donde:

* (SS_{\text{tot}} = \sum (y_i - \bar{y})^2) → **Suma de cuadrados total** (variabilidad total de (Y)).
* (SS_{\text{reg}} = \sum (\hat{y}_i - \bar{y})^2) → **Suma de cuadrados de la regresión** (variabilidad explicada por el modelo).
* (SS_{\text{res}} = \sum (y_i - \hat{y}_i)^2) → **Suma de cuadrados residual** (variabilidad no explicada).

---



## 🔹 Interpretación descriptiva

* (R^2 = 0) → El modelo **no explica nada** de la variabilidad de (Y).
* (R^2 = 1) → El modelo explica el **100% de la variabilidad** de (Y).
* Valores intermedios → Ejemplo: (R^2 = 0.73) significa que el **73% de la variabilidad de (Y)** se debe al modelo y el **27% se debe al azar o a factores no incluidos**.

---

## 🔹 Relación con estadística descriptiva

En esencia, (R^2) **resume una comparación de varianzas**, por lo que se puede ver como una **medida descriptiva de ajuste**:

* Se construye con **medias, varianzas y desviaciones típicas**.
* En el caso de una regresión lineal simple ((Y) frente a (X)), el coeficiente de determinación coincide con el **cuadrado de la correlación de Pearson ((r))**:
  [
  R^2 = r^2
  ]
  Esto muestra que, en regresión lineal simple, (R^2) es una **medida descriptiva de la fuerza de asociación lineal**.

---

## 🔹 Limitaciones

* **No siempre indica causalidad**: un (R^2) alto no implica que (X) cause cambios en (Y).
* **No penaliza complejidad**: más variables siempre tienden a aumentar (R^2), incluso si no aportan valor (por eso existe el (R^2) ajustado).
* Puede ser **engañosa en datos no lineales**: un modelo lineal con datos curvos puede tener bajo (R^2) aunque exista relación.

---

📌 **Ejemplo descriptivo rápido**:
Si estudiamos cómo la **publicidad (X)** explica las **ventas (Y)** y encontramos que (R^2 = 0.82), concluimos que **el 82% de la variación de las ventas se debe al gasto en publicidad**, mientras que el 18% depende de otros factores (competencia, economía, etc.).
