# Funciones y su comportamiento con conjuntos 



Este es un ejercicio clásico que ilustra por qué la imagen de la intersección de dos conjuntos no siempre es igual a la intersección de sus imágenes, una propiedad que solo se cumple bajo ciertas condiciones (como la inyectividad).

A continuación, presento el enunciado formal basado en el estilo de **Bartle & Sherbert** (*Introduction to Real Analysis*), la demostración de la primera parte y la resolución de la pregunta final.

---



## 1. Enunciado del Problema

Sea $f: \mathbb{R} \to \mathbb{R}$ la función definida por $f(x) = x^2$. Sean los conjuntos $E$ y $F$ subconjuntos de los números reales definidos como:
$$
E = \{x \in \mathbb{R} : -1 \le x \le 0\}
$$
$$
F = \{x \in \mathbb{R} : 0 \le x \le 1\}
$$

**Se debe demostrar que:**
1. $E \cap F = \{0\}$ y $f(E \cap F) = \{0\}$.
2. $f(E) = f(F) = \{y \in \mathbb{R} : 0 \le y \le 1\}$.
3. $f(E \cap F)$ es un subconjunto propio de $f(E) \cap f(F)$.

**Pregunta adicional:** ¿Qué ocurre si se suprime el $0$ de los conjuntos $E$ y $F$?

---



## 2. Demostración

### Parte A: Intersección y su imagen
Para hallar $E \cap F$, buscamos los puntos que pertenecen a ambos intervalos:
* $x \in E \iff -1 \le x \le 0$
* $x \in F \iff 0 \le x \le 1$

El único punto que satisface ambas condiciones es $x = 0$. Por lo tanto, **$E \cap F = \{0\}$**.
Aplicando la función:
$$f(E \cap F) = f(\{0\}) = \{0^2\} = \{0\}$$



### Parte B: Imágenes de los conjuntos
Calculamos la imagen de cada conjunto bajo $f(x) = x^2$:
* Para **$E = [-1, 0]$**: Como $x \le 0$, al elevar al cuadrado, los valores se invierten y se vuelven positivos. $0^2 = 0$ y $(-1)^2 = 1$. Por la continuidad de la función, $f(E) = [0, 1]$.
* Para **$F = [0, 1]$**: Claramente, $0^2 = 0$ y $1^2 = 1$. Así, $f(F) = [0, 1]$.

En notación de conjuntos:
$$f(E) = f(F) = \{y \in \mathbb{R} : 0 \le y \le 1\}$$

 and [0,1] on the x-axis and their common image [0,1] on the y-axis]



### Parte C: Subconjunto propio
Calculamos la intersección de las imágenes:
$$f(E) \cap f(F) = [0, 1] \cap [0, 1] = [0, 1]$$

Comparando con el resultado de la Parte A:
* $f(E \cap F) = \{0\}$
* $f(E) \cap f(F) = [0, 1]$

Puesto que $\{0\} \subset [0, 1]$ y $\{0\} \neq [0, 1]$, queda demostrado que **$f(E \cap F)$ es un subconjunto propio de $f(E) \cap f(F)$**.

---



## 3. ¿Qué ocurre si se suprime el 0?

Si definimos los nuevos conjuntos como $E' = E \setminus \{0\} = [-1, 0)$ y $F' = F \setminus \{0\} = (0, 1]$, ocurre lo siguiente:



1.  **La intersección se vuelve vacía:**
    $$E' \cap F' = \emptyset$$
    Consecuentemente, la imagen de la intersección es el conjunto vacío: $f(E' \cap F') = \emptyset$.



2.  **Las imágenes individuales cambian ligeramente:**
    * $f(E') = (0, 1]$
    * $f(F') = (0, 1]$
    *(Notemos que el $0$ ya no está en la imagen porque el único elemento que mapea a $0$ era el propio $0$, el cual fue eliminado).*

3.  **La relación de subconjunto propio se mantiene (y se extrema):**
    $$f(E' \cap F') = \emptyset$$
    $$f(E') \cap f(F') = (0, 1]$$
    Como $\emptyset \subset (0, 1]$, la conclusión de que la imagen de la intersección es un subconjunto propio de la intersección de las imágenes sigue siendo válida, pero ahora los conjuntos son **disjuntos en el dominio** mientras que sus imágenes **coinciden casi totalmente**.

> **Nota de Análisis:** Este fenómeno ocurre porque la función $f(x) = x^2$ no es inyectiva en el conjunto $E \cup F$. Si la función fuera inyectiva, se cumpliría la igualdad $f(E \cap F) = f(E) \cap f(F)$.

# Problema 11  

Este es el ejercicio 11 de la sección de preliminares del **Bartle & Sherbert**. Es la continuación perfecta del anterior porque utiliza el mismo contraejemplo para mostrar que la función imagen tampoco se comporta de forma "intuitiva" con respecto a la diferencia de conjuntos.

Aquí tienes el desarrollo formal siguiendo el rigor del análisis real.

---



## 1. Enunciado del Problema

Sean $f: \mathbb{R} \to \mathbb{R}$ definida por $f(x) = x^2$, y los conjuntos $E = \{x \in \mathbb{R} : -1 \le x \le 0\}$ y $F = \{x \in \mathbb{R} : 0 \le x \le 1\}$.

**Se pide:**
1. Encontrar los conjuntos $E \setminus F$ y $f(E) \setminus f(F)$.
2. Demostrar que la inclusión $f(E \setminus F) \subseteq f(E) \setminus f(F)$ **no es verdadera**.

---



## 2. Solución y Demostración

Recordemos del ejercicio anterior que:
* $E = [-1, 0]$
* $F = [0, 1]$
* $f(E) = [0, 1]$
* $f(F) = [0, 1]$



### Parte A: Determinación de los conjuntos diferencia

**1. Para $E \setminus F$:**
La diferencia de conjuntos $E \setminus F$ consiste en los elementos que están en $E$ pero **no** en $F$.
$$E \setminus F = \{x : -1 \le x \le 0\} \setminus \{x : 0 \le x \le 1\}$$
El único elemento común es el $0$. Al quitarlo de $E$, obtenemos el intervalo semiabierto:
$$E \setminus F = [-1, 0)$$



**2. Para $f(E) \setminus f(F)$:**
Usando los resultados ya obtenidos:
$$f(E) = [0, 1]$$
$$f(F) = [0, 1]$$
Puesto que ambos conjuntos son idénticos, al restarlos obtenemos el conjunto vacío:
$$f(E) \setminus f(F) = \emptyset$$

---



### Parte B: Demostración de la falsedad de la inclusión

Para demostrar que una afirmación de la forma $A \subseteq B$ no es verdadera, basta con encontrar un **contraejemplo**: un elemento que pertenezca a $A$ pero no a $B$.

**Paso 1: Calcular $f(E \setminus F)$**
Sabemos que $E \setminus F = [-1, 0)$. Aplicamos la función $f(x) = x^2$:
* Si $x = -1$, $f(x) = 1$.
* A medida que $x$ se acerca a $0$ por la izquierda, $x^2$ se acerca a $0$, pero nunca llega a ser $0$ porque el $0$ fue excluido del dominio $E \setminus F$.
Por lo tanto:
$$
f(E \setminus F) = (0, 1]
$$





**Paso 2: Comparar los conjuntos**
Tenemos que:
* $f(E \setminus F) = (0, 1]$
* $f(E) \setminus f(F) = \emptyset$

Claramente, $(0, 1] \not\subseteq \emptyset$, ya que el conjunto vacío no contiene elementos, mientras que $(0, 1]$ contiene infinitos puntos (por ejemplo, el $1$ o el $0.5$).

**Conclusión:**
Puesto que existe al menos un elemento (de hecho, todos los del intervalo) que pertenece a $f(E \setminus F)$ pero no a $f(E) \setminus f(F)$, queda demostrado que la inclusión general no se cumple para esta función.



### Nota Teórica (Contexto de Bartle)
En el texto de Bartle, se enseña que para cualquier función $f: A \to B$, se cumple que $f(E) \setminus f(F) \subseteq f(E \setminus F)$. Sin embargo, la inclusión en el sentido opuesto (la que acabamos de invalidar) **solo es verdadera si la función es inyectiva**. 

Como $f(x) = x^2$ no es inyectiva en $E \cup F$ (porque $f(-1) = f(1)$), la igualdad falla estrepitosamente.

# Problema 12  

Este es el **Ejercicio 9** de la sección de preliminares del libro de **Bartle & Sherbert**. Es un teorema fundamental porque establece las reglas de "comportamiento" de las funciones con respecto a las operaciones básicas de conjuntos.

A diferencia de los ejercicios anteriores (que eran contraejemplos específicos), aquí realizaremos una **demostración general** válida para cualquier función y cualquier par de conjuntos.

---



## 1. Enunciado del Problema

Sea $f: A \to B$ una función y sean $E, F$ subconjuntos de $A$. Demostrar las siguientes propiedades de la imagen directa:

1.  **Preservación de la unión:** $f(E \cup F) = f(E) \cup f(F)$
2.  **Inclusión de la intersección:** $f(E \cap F) \subseteq f(E) \cap f(F)$

---



## 2. Demostración

Para demostrar la igualdad de conjuntos en la Parte 1, debemos probar la doble inclusión ($\subseteq$ y $\supseteq$). Para la Parte 2, solo la inclusión simple.

### Parte 1: $f(E \cup F) = f(E) \cup f(F)$

**$(\subseteq)$ Demostramos que $f(E \cup F) \subseteq f(E) \cup f(F)$:**
Sea $y \in f(E \cup F)$. Por definición de imagen directa, existe un $x \in E \cup F$ tal que $f(x) = y$.
* Si $x \in E \cup F$, entonces $x \in E$ o $x \in F$.
* Si $x \in E$, entonces $f(x) \in f(E)$.
* Si $x \in F$, entonces $f(x) \in f(F)$.
En cualquier caso, $y = f(x) \in f(E) \cup f(F)$. Por lo tanto, $f(E \cup F) \subseteq f(E) \cup f(F)$.



**$(\supseteq)$ Demostramos que $f(E) \cup f(F) \subseteq f(E \cup F)$:**
Sea $y \in f(E) \cup f(F)$. Entonces $y \in f(E)$ o $y \in f(F)$.
* Si $y \in f(E)$, existe $x \in E$ tal que $f(x) = y$. Como $E \subseteq E \cup F$, entonces $x \in E \cup F$, lo que implica $y \in f(E \cup F)$.
* Si $y \in f(F)$, existe $x \in F$ tal que $f(x) = y$. Como $F \subseteq E \cup F$, entonces $x \in E \cup F$, lo que implica $y \in f(E \cup F)$.
En ambos casos, $y \in f(E \cup F)$. Por tanto, la inclusión se cumple.

**Conclusión:** Al cumplirse ambas inclusiones, $f(E \cup F) = f(E) \cup f(F)$.



### Parte 2: $f(E \cap F) \subseteq f(E) \cap f(F)$

Sea $y \in f(E \cap F)$.
1. Por definición, existe un elemento $x \in E \cap F$ tal que $f(x) = y$.
2. Como $x \in E \cap F$, entonces $x \in E$ **y** $x \in F$.
3. Dado que $x \in E$, se tiene que $f(x) \in f(E)$.
4. Dado que $x \in F$, se tiene que $f(x) \in f(F)$.
5. Como $y = f(x)$ pertenece a ambos conjuntos, entonces $y \in f(E) \cap f(F)$.

**Conclusión:** Por lo tanto, $f(E \cap F) \subseteq f(E) \cap f(F)$.



## 3. Comentario Crítico (Estilo Bartle)

Es vital notar la diferencia entre ambos resultados. En el primer ejercicio que hicimos ($f(x)=x^2$), vimos que la igualdad en la Parte 2 **no siempre se cumple**. 




> **¿Por qué la intersección no es una igualdad?**
> Para que $y \in f(E) \cap f(F)$, solo se requiere que $y$ tenga *alguna* preimagen en $E$ y *alguna* preimagen en $F$. Estas preimágenes no tienen por qué ser la misma. Sin embargo, para que $y \in f(E \cap F)$, se requiere que exista una **única preimagen común** a ambos conjuntos. Si la función es inyectiva, estas preimágenes están obligadas a ser la misma, y la inclusión se convierte en igualdad.
