Este es el **Ejercicio 1** de la Sección 1.1 de Bartle & Sherbert. Es un ejercicio fundamental para dominar el lenguaje de conjuntos, ya que utiliza el concepto de **bicondicional** ("si y sólo si"). Para probar una equivalencia de este tipo, debemos demostrar la implicación en ambos sentidos.

---



## 1. Enunciado del Problema

Sean $A$ y $B$ dos conjuntos cualesquiera. Demostrar que:
$$A \subseteq B \iff A \cap B = A$$

Donde $\subseteq$ denota la relación de inclusión y $\cap$ la operación de intersección.

---



## 2. Demostración (Estilo Bartle)

Para demostrar el bicondicional $P \iff Q$, procederemos en dos pasos:



### Parte I: Implicación hacia la derecha ($\implies$)
**Supongamos que $A \subseteq B$. Debemos demostrar que $A \cap B = A$.**



Para demostrar la igualdad de conjuntos $A \cap B = A$, debemos probar que cada uno es subconjunto del otro:

1.  **Demostrar que $(A \cap B) \subseteq A$:**
    Por la propia definición de intersección, si $x \in A \cap B$, entonces $x \in A$ y $x \in B$. En particular, $x \in A$. Por lo tanto, $(A \cap B) \subseteq A$ siempre es cierto (esta parte no requiere la hipótesis).

2.  **Demostrar que $A \subseteq (A \cap B)$:**
    Sea $x \in A$. Por nuestra hipótesis original, sabemos que $A \subseteq B$, lo que significa que todo elemento de $A$ también está en $B$. Entonces, $x \in B$.
    Puesto que $x \in A$ y $x \in B$, por definición de intersección, $x \in A \cap B$. Así, $A \subseteq (A \cap B)$.

**Conclusión Parte I:** Al cumplirse ambas inclusiones, $A \cap B = A$.



### Parte II: Implicación hacia la izquierda ($\impliedby$)
**Supongamos ahora que $A \cap B = A$. Debemos demostrar que $A \subseteq B$.**




1.  Sea $x$ un elemento arbitrario tal que $x \in A$.
2.  Por la hipótesis de igualdad, como $A = A \cap B$, cualquier elemento que esté en $A$ debe estar necesariamente en $A \cap B$. Por lo tanto, $x \in A \cap B$.
3.  La definición de intersección establece que si $x \in A \cap B$, entonces $x \in A$ y $x \in B$.
4.  De lo anterior, se sigue directamente que $x \in B$.

**Conclusión Parte II:** Como partimos de un $x \in A$ arbitrario y llegamos a que $x \in B$, queda demostrado que $A \subseteq B$.



## 3. Resumen Lógico

* **En la primera parte**, usamos el hecho de que $A$ está "dentro" de $B$, por lo que la parte común entre ambos es precisamente todo $A$.
* **En la segunda parte**, razonamos que si la parte común entre $A$ y $B$ es el mismo conjunto $A$, es porque no hay nada en $A$ que se quede fuera de $B$.

Esta propiedad es una de las caracterizaciones más útiles de la inclusión en el análisis real.

# Ejercicio 2 de la sección 1.1  

Este es el **Ejercicio 2** de la sección 1.1 del libro de **Bartle & Sherbert**. Se nos pide demostrar una de las leyes fundamentales de la teoría de conjuntos, la cual establece cómo interactúa el complemento con la intersección.

---



## 1. Enunciado del Problema

Sea $S$ un conjunto universal y sean $A, B \subseteq S$. Demostrar la **segunda ley de De Morgan**, que se enuncia como:
$$A \setminus (B \cap C) = (A \setminus B) \cup (A \setminus C)$$

En el contexto general de complementos (si $A$ es el conjunto universal), la ley se suele escribir como:
$$(B \cap C)^c = B^c \cup C^c$$

> **Nota:** El texto de Bartle utiliza la notación de diferencia de conjuntos $A \setminus B$ para referirse a los elementos que están en $A$ pero no en $B$.

---



## 2. Demostración

Para demostrar la igualdad de dos conjuntos, debemos demostrar la doble inclusión: $\subseteq$ y $\supseteq$.


### Parte I: $A \setminus (B \cap C) \subseteq (A \setminus B) \cup (A \setminus C)$

1.  Sea $x \in A \setminus (B \cap C)$.
2.  Por definición de diferencia de conjuntos, esto significa que $x \in A$ **y** $x \notin (B \cap C)$.
3.  La afirmación $x \notin (B \cap C)$ es la negación lógica de "$x$ está en $B$ y $x$ está en $C$". Según las reglas de la lógica (negación de una conjunción), esto significa que:
    * $x \notin B$ **o** $x \notin C$.
4.  Ahora distribuimos la condición $x \in A$:
    * $(x \in A \text{ y } x \notin B)$ **o** $(x \in A \text{ y } x \notin C)$.
5.  Traduciendo esto de nuevo a notación de conjuntos:
    * $x \in (A \setminus B)$ **o** $x \in (A \setminus C)$.
6.  Por definición de unión, $x \in (A \setminus B) \cup (A \setminus C)$.

---



### Parte II: $(A \setminus B) \cup (A \setminus C) \subseteq A \setminus (B \cap C)$

1.  Sea $x \in (A \setminus B) \cup (A \setminus C)$.
2.  Por definición de unión, $x \in (A \setminus B)$ **o** $x \in (A \setminus C)$.
3.  Esto implica que:
    * $(x \in A \text{ y } x \notin B)$ **o** $(x \in A \text{ y } x \notin C)$.
4.  En ambos casos, observamos que $x \in A$ es una condición común. Por lo tanto, $x \in A$ **y** ($x \notin B$ o $x \notin C$).
5.  La expresión ($x \notin B$ o $x \notin C$) es equivalente a decir que $x$ no puede estar en la intersección de ambos, es decir, $x \notin (B \cap C)$.
6.  Entonces, tenemos que $x \in A$ **y** $x \notin (B \cap C)$.
7.  Por definición de diferencia, $x \in A \setminus (B \cap C)$.

---



## 3. Conclusión

Puesto que hemos demostrado que cada conjunto está contenido en el otro, concluimos que:
$$A \setminus (B \cap C) = (A \setminus B) \cup (A \setminus C)$$

Esta ley nos dice, intuitivamente, que para quedar fuera de una "zona común" (intersección), basta con quedar fuera de al menos uno de los conjuntos que la componen. Es una herramienta esencial en el análisis real para trabajar con vecindades y conjuntos abiertos/cerrados.

# Ejercicio 3 de la sección 1.1  

Este es el **Ejercicio 3** de la sección 1.1 del libro de **Bartle & Sherbert**. Estas leyes son los pilares algebraicos de la teoría de conjuntos; permiten "distribuir" una operación sobre otra, de manera similar a como distribuimos la multiplicación sobre la suma en los números reales.

---

## 1. Enunciado del Problema

Sean $A, B$ y $C$ subconjuntos de un conjunto universal $S$. Demostrar las **leyes distributivas**:

* **a)** $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$
* **b)** $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$

---

## 2. Demostración de la ley (a)
Queremos probar que $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$.


### $(\subseteq)$ Inclusión hacia la derecha:
1.  Sea $x \in A \cap (B \cup C)$.
2.  Por definición de intersección, $x \in A$ **y** $x \in (B \cup C)$.
3.  De $x \in (B \cup C)$, sabemos que $x \in B$ **o** $x \in C$.
4.  Si $x \in B$, como también sabemos que $x \in A$, entonces $x \in (A \cap B)$.
5.  Si $x \in C$, como también sabemos que $x \in A$, entonces $x \in (A \cap C)$.
6.  En cualquiera de los dos casos, $x \in (A \cap B) \cup (A \cap C)$.

### $(\supseteq)$ Inclusión hacia la izquierda:
1.  Sea $x \in (A \cap B) \cup (A \cap C)$.
2.  Entonces $x \in (A \cap B)$ **o** $x \in (A \cap C)$.
3.  En ambos casos, $x \in A$.
4.  Además, en el primer caso $x \in B$ y en el segundo $x \in C$. Por lo tanto, $x \in (B \cup C)$.
5.  Como $x \in A$ y $x \in (B \cup C)$, entonces $x \in A \cap (B \cup C)$.

---

## 3. Demostración de la ley (b)
Queremos probar que $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$.


### $(\subseteq)$ Inclusión hacia la derecha:
1.  Sea $x \in A \cup (B \cap C)$.
2.  Esto significa que $x \in A$ **o** $x \in (B \cap C)$.
3.  **Caso 1:** Si $x \in A$, entonces $x \in (A \cup B)$ y también $x \in (A \cup C)$. Por lo tanto, $x \in (A \cup B) \cap (A \cup C)$.
4.  **Caso 2:** Si $x \in (B \cap C)$, entonces $x \in B$ y $x \in C$.
    * Como $x \in B$, entonces $x \in (A \cup B)$.
    * Como $x \in C$, entonces $x \in (A \cup C)$.
    * Luego, $x \in (A \cup B) \cap (A \cup C)$.
5.  En ambos casos, la inclusión se cumple.

### $(\supseteq)$ Inclusión hacia la izquierda:
1.  Sea $x \in (A \cup B) \cap (A \cup C)$.
2.  Entonces $x \in (A \cup B)$ **y** $x \in (A \cup C)$.
3.  Si $x \in A$, entonces claramente $x \in A \cup (B \cap C)$ y terminamos.
4.  Si $x \notin A$, entonces para que se cumplan las condiciones del punto 2:
    * De $x \in (A \cup B)$ se deduce que $x$ debe estar en $B$.
    * De $x \in (A \cup C)$ se deduce que $x$ debe estar en $C$.
5.  Por lo tanto, $x \in (B \cap C)$, lo que implica que $x \in A \cup (B \cap C)$.

---



## 4. Conclusión Teórica
Estas leyes demuestran que la unión y la intersección son "duales" entre sí. A diferencia del álgebra de números reales (donde la suma no se distribuye sobre la multiplicación), en el álgebra de conjuntos de **Bartle**, ambas operaciones son completamente distributivas la una respecto a la otra.
