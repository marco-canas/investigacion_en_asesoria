# 1. Demostración: $A \subseteq B \iff A \cap B = A$

Debemos demostrar una **doble implicación**:

$$
A \subseteq B \quad \Longleftrightarrow \quad A \cap B = A
$$

Por lo tanto demostramos dos partes.

---



## (⇒) Si $A \subseteq B$, entonces $A \cap B = A$

Supongamos que

$$
A \subseteq B
$$

Esto significa que

$$
\forall x (x \in A \Rightarrow x \in B)
$$



Ahora tomemos un elemento $x$ en la intersección:

$$
x \in A \cap B
$$

Por definición de intersección:

$$
x \in A \quad \text{y} \quad x \in B
$$



En particular

$$
x \in A
$$

Entonces

$$
A \cap B \subseteq A
$$

Ahora tomemos $x \in A$.

Como $A \subseteq B$, entonces

$$
x \in B
$$

por lo tanto

$$
x \in A \cap B
$$

Esto implica

$$
A \subseteq A \cap B
$$

Por lo tanto

$$
A \cap B = A
$$

---

## (⇐) Si $A \cap B = A$, entonces $A \subseteq B$

Supongamos que

$$
A \cap B = A
$$

Tomemos cualquier elemento

$$
x \in A
$$

Como

$$
A = A \cap B
$$

entonces

$$
x \in A \cap B
$$

Por definición de intersección

$$
x \in A \quad \text{y} \quad x \in B
$$

Luego

$$
x \in B
$$

Por lo tanto

$$
A \subseteq B
$$

---

✅ Concluimos

$$
A \subseteq B \iff A \cap B = A
$$

---

# Exploración computacional con Python

Podemos **experimentar con conjuntos concretos**.

## Código con pandas

```python
import pandas as pd

A = {1,2,3}
B = {1,2,3,4,5}

interseccion = A.intersection(B)

df = pd.DataFrame({
    "A": list(A),
    "A ∩ B": list(interseccion)
})

print("A ⊆ B:", A.issubset(B))
print("A ∩ B =", interseccion)

df
```

Resultado esperado:

```
A ⊆ B: True
A ∩ B = {1,2,3}
```

Lo cual confirma que

$$
A \cap B = A
$$

# Visualización con matplotlib (diagrama tipo Venn)

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

fig, ax = plt.subplots()

A_circle = Circle((0.4,0.5),0.3,alpha=0.3,label='A')
B_circle = Circle((0.6,0.5),0.4,alpha=0.3,label='B')

ax.add_patch(A_circle)
ax.add_patch(B_circle)

ax.set_aspect('equal')
plt.title("Caso A ⊆ B")
plt.show()
```

Interpretación:

Cuando

$$
A \subseteq B
$$

el conjunto $A$ está **contenido dentro de $B$** y por eso

$$
A \cap B = A
$$

# 2. Segunda Ley de De Morgan

Debemos demostrar

$$
(A \cap B)^c = A^c \cup B^c
$$

---

## Demostración

Tomemos

$$
x \in (A \cap B)^c
$$

Esto significa

$$
x \notin A \cap B
$$

La definición de intersección dice

$$
x \in A \cap B \iff x \in A \ \text{y} \ x \in B
$$

Por lo tanto

$$
x \notin A \cap B
$$

significa

$$
x \notin A \quad \text{o} \quad x \notin B
$$

lo cual equivale a

$$
x \in A^c \quad \text{o} \quad x \in B^c
$$

Por definición de unión

$$
x \in A^c \cup B^c
$$

Por lo tanto

$$
(A \cap B)^c \subseteq A^c \cup B^c
$$

---

Ahora tomemos

$$
x \in A^c \cup B^c
$$

Entonces

$$
x \in A^c \quad \text{o} \quad x \in B^c
$$

lo que significa

$$
x \notin A \quad \text{o} \quad x \notin B
$$

Por lo tanto

$$
x \notin A \cap B
$$

es decir

$$
x \in (A \cap B)^c
$$

Así

$$
A^c \cup B^c \subseteq (A \cap B)^c
$$

---

Finalmente

$$
(A \cap B)^c = A^c \cup B^c
$$

---

# Verificación computacional

```python
U = set(range(1,11))

A = {1,2,3,4}
B = {3,4,5,6}

inter = A.intersection(B)

left = U - inter
right = (U-A).union(U-B)

print("(A ∩ B)^c =", left)
print("A^c ∪ B^c =", right)

print("Son iguales:", left == right)
```

---

# Interpretación visual

Geométricamente:

* $A \cap B$ = región común
* su complemento = todo **excepto esa región**

Eso coincide exactamente con

$$
A^c \cup B^c
$$
