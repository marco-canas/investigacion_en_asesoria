Vamos a resolver el ejercicio paso a paso.  

El enunciado dice:  
Sean los números \( x_n \) definidos por  
\[
x_1 := 1, \quad x_2 := 2, \quad x_{n+2} := \frac{1}{2}(x_{n+1} + x_n) \quad \text{para toda } n \in \mathbb{N}.
\]  
Queremos demostrar, por inducción fuerte, que \( 1 \leq x_n \leq 2 \) para toda \( n \in \mathbb{N} \).

---

## 1. Demostración por inducción fuerte

**Base de inducción**:  
Para \( n = 1 \): \( x_1 = 1 \Rightarrow 1 \leq 1 \leq 2 \), se cumple.  
Para \( n = 2 \): \( x_2 = 2 \Rightarrow 1 \leq 2 \leq 2 \), se cumple.

**Hipótesis de inducción fuerte**:  
Supongamos que para algún \( k \geq 2 \) se cumple que \( 1 \leq x_j \leq 2 \) para todo \( j \leq k \).

**Paso inductivo**:  
Queremos demostrar que \( 1 \leq x_{k+1} \leq 2 \).  

Si \( k+1 > 2 \), entonces \( k+1 \geq 3 \) y podemos escribir:
\[
x_{k+1} = \frac{1}{2} (x_k + x_{k-1}).
\]
Por hipótesis de inducción:
\[
1 \leq x_k \leq 2, \quad 1 \leq x_{k-1} \leq 2.
\]
Sumando:
\[
2 \leq x_k + x_{k-1} \leq 4.
\]
Dividiendo entre 2:
\[
1 \leq \frac{x_k + x_{k-1}}{2} \leq 2.
\]
Por lo tanto, \( 1 \leq x_{k+1} \leq 2 \), como se quería demostrar.

Queda probado por inducción fuerte que para todo \( n \in \mathbb{N} \), \( 1 \leq x_n \leq 2 \).

---

## 2. Cálculo de los primeros términos y visualización con Python

Podemos generar los primeros términos de la sucesión y ver que efectivamente están en \([1, 2]\).

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np

# Definimos la sucesión
def generar_xn(n_max):
    x = [1, 2]  # x1, x2
    for i in range(2, n_max):
        x.append(0.5 * (x[-1] + x[-2]))
    return x

# Generamos hasta n=20
n_terminos = 20
x_vals = generar_xn(n_terminos)

# Mostramos en tabla
df = pd.DataFrame({
    'n': range(1, n_terminos + 1),
    'x_n': x_vals
})
print(df)

# Graficamos con matplotlib
plt.figure(figsize=(10, 5))
plt.plot(range(1, n_terminos + 1), x_vals, marker='o', linestyle='-', color='b')
plt.axhline(y=1, color='r', linestyle='--', label='Cota inferior = 1')
plt.axhline(y=2, color='g', linestyle='--', label='Cota superior = 2')
plt.title('Sucesión $x_n$ con $x_{n+2} = \\frac{1}{2}(x_{n+1}+x_n)$')
plt.xlabel('n')
plt.ylabel('$x_n$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico interactivo con plotly
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=list(range(1, n_terminos + 1)),
    y=x_vals,
    mode='lines+markers',
    name='$x_n$'
))
fig.add_hline(y=1, line_dash="dash", line_color="red", annotation_text="Cota inferior=1")
fig.add_hline(y=2, line_dash="dash", line_color="green", annotation_text="Cota superior=2")
fig.update_layout(
    title="Sucesión $x_n$ con $x_{n+2} = \\frac{1}{2}(x_{n+1}+x_n)$",
    xaxis_title="n",
    yaxis_title="$x_n$"
)
fig.show()
```

## 3. Interpretación de resultados

En la tabla y gráficas se observa que \( x_n \) siempre está entre 1 y 2, como se demostró formalmente.  
Además, la sucesión tiende a un valor límite. Podemos calcular el límite:  
Si suponemos que \( \lim_{n \to \infty} x_n = L \), entonces  
\[
L = \frac{1}{2}(L + L) \ \text{(trivial)}.
\]
Usando la relación de recurrencia, tomando límite en la ecuación \( x_{n+2} = \frac12(x_{n+1} + x_n) \) obtenemos \( L = \frac12(L + L) \) que no da información nueva.  
Resolvamos la recurrencia lineal homogénea:  
\[
x_{n+2} - \frac12 x_{n+1} - \frac12 x_n = 0.
\]
Ecuación característica: \( r^2 - \frac12 r - \frac12 = 0 \),  
multiplicamos por 2: \( 2r^2 - r - 1 = 0 \),  
soluciones: \( r = 1 \) y \( r = -\frac12 \).  
Solución general:  
\[
x_n = A \cdot 1^n + B \cdot \left(-\frac12\right)^n.
\]
Con \( x_1 = 1 \) y \( x_2 = 2 \):  
\[
A - \frac{B}{2} = 1, \quad A + \frac{B}{4} = 2.
\]
Restando: \( \frac{3B}{4} = 1 \Rightarrow B = \frac{4}{3} \).  
Luego \( A = 1 + \frac12 \cdot \frac{4}{3} = 1 + \frac{2}{3} = \frac{5}{3} \).  
Entonces:
\[
x_n = \frac{5}{3} + \frac{4}{3} \left(-\frac12\right)^n.
\]
Claramente \( \lim_{n \to \infty} x_n = \frac{5}{3} \approx 1.666\ldots \), que está entre 1 y 2.
