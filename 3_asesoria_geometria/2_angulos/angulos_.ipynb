La diferencia de las medidas de dos ángulos adyacentes AOB y BOC es $30^{o}$. Calcule la medida del ángulo formado por la bisectriz del ángulo AOC y el lado OB. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Condición del problema
AOB = 80
BOC = 50     # diferencia = 30°

# Direcciones de los rayos
theta_OB = 0
theta_OA = AOB
theta_OC = -BOC

# Bisectriz del ángulo AOC
theta_bis = (theta_OA + theta_OC) / 2

r = 5

def punto(ang):
    ang = np.deg2rad(ang)
    return r*np.cos(ang), r*np.sin(ang)

O = (0,0)
A = punto(theta_OA)
B = punto(theta_OB)
C = punto(theta_OC)
D = punto(theta_bis)

plt.figure(figsize=(6,6))

# Rayos
plt.plot([0,A[0]],[0,A[1]], linewidth=2,label="OA")
plt.plot([0,B[0]],[0,B[1]], linewidth=2,label="OB")
plt.plot([0,C[0]],[0,C[1]], linewidth=2,label="OC")
plt.plot([0,D[0]],[0,D[1]], '--', linewidth=2,label="Bisectriz AOC")

# Etiquetas
plt.text(*A," A")
plt.text(*B," B")
plt.text(*C," C")
plt.text(0,0,"O")

# Arco AOB
t = np.linspace(np.deg2rad(theta_OB),np.deg2rad(theta_OA),100)
plt.plot(1.2*np.cos(t),1.2*np.sin(t))

# Arco BOC
t = np.linspace(np.deg2rad(theta_OC),np.deg2rad(theta_OB),100)
plt.plot(1.6*np.cos(t),1.6*np.sin(t))

# Arco AOC
t = np.linspace(np.deg2rad(theta_OC),np.deg2rad(theta_OA),100)
plt.plot(2*np.cos(t),2*np.sin(t))

plt.axhline(0)
plt.axvline(0)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.legend()

plt.title("Bisectriz del ángulo AOC")

plt.show()

# Ángulo pedido
angulo = abs(theta_bis - theta_OB)
print("Ángulo entre la bisectriz y OB:", angulo)
```
