# Problemas de ángulos (Geometría Euclidiana)

### Problema 1

La suma de los complementos de dos ángulos es 130° y la diferencia de sus suplementos es 10°. Determine las medidas de los ángulos.

### Problema 2

Dos ángulos son complementarios. Uno mide 18° más que el otro. Determine la medida de cada ángulo.

### Problema 3

Dos ángulos son suplementarios. Uno mide el triple del otro. Determine sus medidas.

### Problema 4

La suma de dos ángulos es 150°. Uno de ellos excede al otro en 30°. Determine ambos ángulos.

### Problema 5

El suplemento de un ángulo es el doble de su complemento. Halle el ángulo.

### Problema 6

Dos ángulos suman 180°. El mayor excede al menor en 40°. Determine sus medidas.

### Problema 7

El complemento de un ángulo es igual a la mitad de su suplemento. Determine el ángulo.

### Problema 8

Dos ángulos son tales que el doble del primero más el segundo suman 210°. Además, el segundo es 30° menor que el primero. Determine ambos ángulos.

### Problema 9

La suma de tres ángulos es 180°. El segundo mide el doble del primero y el tercero mide 20° más que el primero. Determine los tres ángulos.

### Problema 10

El suplemento de un ángulo excede su complemento en 60°. Determine el ángulo.

---



# Problemas con rectas y ángulos

### Problema 11

Dos ángulos adyacentes forman un ángulo llano. Uno mide 35° más que el otro. Determine sus medidas.

### Problema 12

Dos rectas se intersectan formando ángulos opuestos por el vértice. Uno de los ángulos mide el triple de uno de sus adyacentes. Determine las medidas de los cuatro ángulos.

### Problema 13

Dos rectas paralelas son cortadas por una transversal. Un ángulo interior mide 25° más que su alterno interno correspondiente. Determine las medidas de los ángulos.

### Problema 14

En una intersección de dos rectas, la suma de dos ángulos adyacentes es cinco veces uno de ellos. Determine las medidas de todos los ángulos.

### Problema 15

Dos rectas paralelas son cortadas por una transversal. La diferencia entre un ángulo interior y su correspondiente es 20°. Determine sus medidas.

---



# Problemas con triángulos

### Problema 16

En un triángulo, un ángulo mide el doble de otro y el tercero mide 30° más que el menor. Determine los ángulos del triángulo.

### Problema 17

En un triángulo isósceles, el ángulo del vértice mide 40° menos que la suma de los ángulos de la base. Determine los tres ángulos.

### Problema 18

En un triángulo, el mayor ángulo excede al menor en 50°, y el tercer ángulo es el promedio de los otros dos. Determine los ángulos.

### Problema 19

En un triángulo rectángulo, uno de los ángulos agudos mide 20° más que el otro. Determine las medidas de los ángulos.

### Problema 20

La suma de dos ángulos exteriores de un triángulo, tomados en vértices diferentes, es 200°. Determine el tercer ángulo exterior y los ángulos interiores del triángulo.
