```python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

```

# 5.3. Propiedades del Determinante

En esta sección se demuestran dos propiedades fundamentales del determinante. La primera establece que el determinante del producto de dos matrices es el producto de sus determinantes. La segunda afirma que el determinante de la inversa de una matriz invertible es el inverso multiplicativo del determinante de la matriz.

Además, se obtiene una nueva caracterización de matrices invertibles, que complementa las caracterizaciones presentadas anteriormente en el Corolario 1.22.

---

## Teorema 5.17

Sean $A$ y $B$ matrices $n\times n$. Entonces

$$
\det(AB)=\det(A)\det(B).
$$

### Demostración

**Caso 1.** Supongamos que $A$ es invertible. Por el Teorema 1.17 existe una factorización

$$
A=E_1E_2\cdots E_k
$$

donde $E_1,\dots,E_k$ son matrices elementales. Entonces

$$
AB=E_1E_2\cdots E_kB.
$$

Usando el Corolario 5.16 repetidamente obtenemos

$$
\det(AB)=\det(E_1)\cdots\det(E_k)\det(B).
$$

Por otra parte,

$$
\det(A)=\det(E_1\cdots E_k)=\det(E_1)\cdots\det(E_k).
$$

Por tanto,

$$
\det(AB)=\det(A)\det(B).
$$

---

**Caso 2.** Supongamos ahora que $A$ no es invertible. Por el Teorema 1.15 existe una factorización

$$
A=E_1\cdots E_kA'
$$

donde $E_1,\dots,E_k$ son matrices elementales y $A'$ está en forma escalonada reducida.

Como $A$ no es invertible, se tiene

$$
\text{rango}(A)=\text{rango}(A')<n.
$$

Como $A'$ es cuadrada, debe tener al menos una fila de ceros. Por tanto, la matriz $A'B$ también tiene una fila de ceros y entonces

$$
\det(A'B)=0.
$$

Usando nuevamente el Corolario 5.16 obtenemos

$$
\det(AB)
========

# \det(E_1\cdots E_kA'B)

# \det(E_1)\cdots\det(E_k)\det(A'B)

0.

$$

Por otra parte,

$$
\det(A)
=======

# \det(E_1\cdots E_kA')

\det(E_1)\cdots\det(E_k)\det(A').
$$

Como $A'$ tiene una fila nula, $\det(A')=0$, y por tanto

$$
\det(A)\det(B)=0.
$$

Concluimos que

$$
\det(AB)=\det(A)\det(B).
$$

---

## Teorema 5.18 (Caracterización de matrices invertibles)

Sea $A\in M_n(\mathbb{R})$. Entonces $A$ es invertible si y sólo si

$$
\det(A)\neq 0.
$$

### Demostración

**$(\Rightarrow)$**
Supongamos que $A$ es invertible. Entonces

$$
A=E_1E_2\cdots E_k
$$

donde $E_1,\dots,E_k$ son matrices elementales. Como cada matriz elemental tiene determinante distinto de cero, se tiene

$$
\det(A)=\det(E_1)\cdots\det(E_k)\neq0.
$$

---

**$(\Leftarrow)$**
Supongamos que $\det(A)\neq0$. Si $A$ no fuera invertible, entonces al llevarla a forma escalonada reducida se obtendría una matriz $A'$ con una fila de ceros. Pero entonces

$$
\det(A')=0
$$

y por la propiedad multiplicativa del determinante se tendría

$$
\det(A)=\det(E_1\cdots E_kA')=0,
$$

lo cual contradice la hipótesis. Por lo tanto $A$ es invertible.

---

## Corolario 5.19

Si $A\in M_n(\mathbb{R})$ es invertible entonces

$$
\det(A^{-1})=\frac{1}{\det(A)}.
$$

### Demostración

Por el Teorema 5.17,

$$
\det(A)\det(A^{-1})
===================

# \det(AA^{-1})

# \det(I)

1.

$$

Por tanto,

$$
\det(A^{-1})=\frac{1}{\det(A)}.
$$

# 5.4 Matriz Adjunta

En esta sección se define la **matriz adjunta** de una matriz cuadrada. La motivación principal es que permite expresar la inversa de una matriz invertible mediante una fórmula explícita.

Si $A$ no es invertible, veremos que el producto de $A$ con su adjunta es la matriz nula.

---



## Definición 5.4

Sea $A\in M_n(\mathbb{R})$. Definimos la **matriz adjunta** de $A$, denotada por $\mathrm{Adj}(A)$, como la traspuesta de la matriz de cofactores:

$$
\mathrm{Adj}(A)=
\begin{bmatrix}
C_{11}&C_{12}&\cdots&C_{1n}\\ 
C_{21}&C_{22}&\cdots&C_{2n}\\ 
\vdots&\vdots&\ddots&\vdots\\ 
C_{n1}&C_{n2}&\cdots&C_{nn}
\end{bmatrix}^t
=
\begin{bmatrix}
C_{11}&C_{21}&\cdots&C_{n1}\\ 
C_{12}&C_{22}&\cdots&C_{n2}\\ 
\vdots&\vdots&\ddots&\vdots\\ 
C_{1n}&C_{2n}&\cdots&C_{nn}
\end{bmatrix}
$$

donde

$$
C_{ij}=(-1)^{i+j}\det(A_{ij})
$$

es el $ij$-ésimo cofactor de $A$.

---



## Teorema 5.21

Si $A\in M_n(\mathbb{R})$ entonces

$$
A,\mathrm{Adj}(A)=\det(A),I.
$$

De este resultado se deduce que:

1. Si $A$ es invertible entonces

$$
A^{-1}=\frac{1}{\det(A)},\mathrm{Adj}(A).
$$

2. Si $A$ no es invertible entonces

$$
A,\mathrm{Adj}(A)=0.
$$

# 5.5. La Regla de Cramer

En esta breve sección se presenta el resultado conocido como **Regla de Cramer**. Este método proporciona una fórmula explícita para calcular la solución de un sistema de ecuaciones lineales cuando dicha solución es única. Si el sistema no tiene solución única, entonces este método no puede aplicarse.



Desde el punto de vista computacional, la regla de Cramer es ineficiente para sistemas grandes, por lo que en la práctica rara vez se utiliza para resolver sistemas lineales. Sin embargo, puede ser útil cuando sólo se requiere calcular una de las incógnitas del sistema, como ocurre en el Problema 5.5.2.

Incluimos esta sección principalmente para demostrar el resultado y para los lectores interesados en esta técnica clásica.

---



## Teorema 5.22 (Regla de Cramer)

Sea $Ax=b$ un sistema de ecuaciones lineales donde $A$ es una matriz $n\times n$ invertible. Sean $A_1,\dots,A_n$ las columnas de $A$ y, para cada $i=1,\dots,n$, definimos $B_i$ como la matriz que se obtiene de $A$ al reemplazar su $i$-ésima columna por el vector $b$. Es decir,

$$
B_1=[, b; A_2; \cdots; A_n ,],
$$

$$
B_2=[, A_1; b; \cdots; A_n ,],
$$

$$
\vdots
$$

$$
B_n=[, A_1; A_2; \cdots; b ,].
$$

Entonces la solución única del sistema está dada por

$$
x_1=\frac{\det(B_1)}{\det(A)}, \quad
x_2=\frac{\det(B_2)}{\det(A)}, \quad \dots \quad
x_n=\frac{\det(B_n)}{\det(A)}.
$$

---



### Demostración

Como $A$ es invertible, el sistema tiene una única solución

$$
x=
\begin{bmatrix}
x_1\\ 
\vdots\\ 
x_n
\end{bmatrix},
$$

lo cual se deduce del Teorema 1.12.



Entonces

$$
b=Ax.
$$



Si $A_1,\dots,A_n$ son las columnas de $A$, se tiene

$$
Ax=
[,A_1; \cdots; A_n,]
\begin{bmatrix}
x_1\\ 
\vdots\\ 
x_n
\end{bmatrix}
=
x_1A_1+\cdots+x_nA_n.
$$

Ahora, para un índice fijo $i$ con $1\le i\le n$, usando las propiedades del determinante con respecto a las columnas (Problema 5.2.9), obtenemos

$$
\det(B_i)
=
\det(A_1,\dots,b,\dots,A_n).
$$



Sustituyendo $b=x_1A_1+\cdots+x_nA_n$,

$$
\det(B_i)
=
\det(A_1,\dots,x_1A_1+\cdots+x_nA_n,\dots,A_n).
$$

Usando la linealidad del determinante en la columna $i$,

$$
\det(B_i)
=
x_1\det(A_1,\dots,A_1,\dots,A_n)
+\cdots+
x_i\det(A_1,\dots,A_i,\dots,A_n)
+\cdots+
x_n\det(A_1,\dots,A_n,\dots,A_n).
$$

Todos los determinantes con columnas repetidas son cero. Por tanto,

$$
\det(B_i)=x_i\det(A).
$$

Despejando $x_i$ obtenemos

$$
x_i=\frac{\det(B_i)}{\det(A)}.
$$

# Ejemplo 5.11

Resuelva el sistema

$$
3x+2y-z=0
$$

$$
-x+3z=0
$$

$$
x+z=1
$$

usando la regla de Cramer.

---



### Solución

Primero escribimos el sistema en forma matricial $Ax=b$, donde

$$
A=
\begin{bmatrix}
3&2&-1\\ 
-1&0&3\\ 
1&0&1
\end{bmatrix},
\quad
b=
\begin{bmatrix}
0\\ 
0\\ 
1
\end{bmatrix}.
$$

Calculamos los determinantes necesarios.

---



$$
\det(B_1)=
\det
\begin{bmatrix}
0&2&-1\\ 
0&0&3\\ 
1&0&1
\end{bmatrix}
=
1
\det
\begin{bmatrix}
2&-1\\ 
0&3
\end{bmatrix}
=6
$$

---



$$
\det(B_2)=
\det
\begin{bmatrix}
3&0&-1\
-1&0&3\
1&1&1
\end{bmatrix}
=
-1
\det
\begin{bmatrix}
3&-1\
-1&3
\end{bmatrix}
=-8
$$

---



$$
\det(B_3)=
\det
\begin{bmatrix}
3&2&0\\ 
-1&0&0\\ 
1&0&1
\end{bmatrix}
=
-2
\det
\begin{bmatrix}
-1&0\\ 
1&1
\end{bmatrix}
=2
$$

---



Ahora calculamos

$$
\det(A)=
\det
\begin{bmatrix}
3&2&-1\\ 
-1&0&3\\
1&0&1
\end{bmatrix}
=
-2
\det
\begin{bmatrix}
-1&3\\ 
1&1
\end{bmatrix}
=8.
$$



Por lo tanto, por la regla de Cramer,

$$
x=\frac{\det(B_1)}{\det(A)}=\frac{6}{8}=\frac{3}{4},
$$

$$
y=\frac{\det(B_2)}{\det(A)}=\frac{-8}{8}=-1,
$$

$$
z=\frac{\det(B_3)}{\det(A)}=\frac{2}{8}=\frac{1}{4}.
$$
